#2 債券

##2-1 債券の基礎

### 債券とは
- 投資家が企業や政府にお金を貸すことで成立する「負債投資」。
- 発行体は一定期間資金を借り、利子を支払う。
- 「固定利付証券（Fixed Income Securities）」とも呼ばれる。

---

### 債券の種類

#### 1. ゼロクーポン債（Zero Coupon Bond）
- 元本のみ返済、利子なし。
- 価格（現在価値）:
  $$P = \frac{X}{(1+r)^N}$$
- 例：額面 1000ドル、満期 2年、市場金利 4% 
  → 価格  ≒ $924

#### 2. クーポン債（Coupon Bond）
- 毎年クーポン（利子）を支払い、満期時に元本も返済。
- 価格（現在価値）:
  $$P = \sum_{t=1}^N \frac{C}{(1+r)^t} + \frac{X}{(1+r)^N}$$
  または  
  $$P = \frac{C}{r}\left(1 - \frac{1}{(1+r)^N}\right) + \frac{X}{(1+r)^N}$$
- 例：額面 1000ドル、クーポン率 10%、満期 3年、市場金利4% 
  → 価格 ≒ $1167

---

👉 債券価格は「将来キャッシュフローの現在価値の合計」で決まる。  
👉 計算時は **市場金利** を使う点が重要。


##2-2 債券の利回りと最終利回り（YTM）

### 利回り（Yield）
- 投資収益を示す指標。
- 計算式：
  $$\text{Yield} = \frac{\text{年間クーポン支払額}}{\text{債券価格}}$$
- 例：価格 $1,167、年間クーポン $100 → Yield ≒ 8.6%

---

### 最終利回り（Yield to Maturity, YTM）
- 債券を満期まで保有した場合の「実質利回り」。
- 内部収益率(IRR) に相当。
- 定義：  
  $$\text{債券価格} = \sum_{t=1}^N \frac{C}{(1+y)^t} + \frac{X}{(1+y)^N}$$
- 連続モデルでは：  
  $$PV = C \cdot e^{-yt}, \quad PV = X \cdot e^{-yT}$$
- 数値解法（ニュートン法など）で $(y)$ を求める。

---

### イールドカーブ（Yield Curve）
- 縦軸：利回り、横軸：満期（年数）。
- 通常は右上がり。
- 理由：長期債ほどリスク大 → 高い利回りを要求。
- リスク要因：
  - 金利変動リスク
  - 経済変動リスク

---

👉 Yield = 単純利回り、YTM = 実際の内部収益率。  
👉 イールドカーブは「期間が長いほど利回りが高い」傾向を示す。


##2-3 市場金利と債券価格の関係

### 基本原則
- 債券価格と市場金利は **逆相関**。
  - 金利 ↑ → 債券価格 ↓
  - 金利 ↓ → 債券価格 ↑

---

### ゼロクーポン債
- 価格：
  $$P = \frac{X}{(1+R)^N}$$
- R（市場金利）が大きいほど価格は小さくなる。

---

### クーポン債
- 価格：
  $$P = \sum_{t=1}^N \frac{C}{(1+R)^t} + \frac{X}{(1+R)^N}$$
- 分母に金利があるため、金利上昇 → 債券価格下落。

---

### 金利リスク（Interest Rate Risk）
- 金利上昇は債券価格を押し下げ、損失リスクを生む。
- 金利低下は債券価格を押し上げ、利益につながる。
- 投資家は将来の金利動向を分析する必要がある。

---

👉 債券価格は「市場金利と逆相関」する。  
👉 これが金利リスクの理論的背景。


##2-4 マコーレー・デュレーション（Macaulay Duration）

### 定義
- 債券のキャッシュフローで投資元本が回収されるまでの平均的な時間。
- 数式：
  $$D = -\frac{1}{V}\cdot \frac{dV}{dy}$$
  - V = 債券価格
  - y = 利回り（YTM）

---

### 特徴
- **ゼロクーポン債**：デュレーション = 満期までの期間。
- **クーポン債**：クーポンと元本を現在価値で加重平均した期間。

---

### 金利リスクとの関係
- 債券価格と金利は逆相関。
- 長期債（デュレーション大）：価格は金利変動に敏感。
- 短期債（デュレーション小）：価格は金利変動に鈍感。

---

### 投資家の行動
- 金利↓ 予想 → 長期債を選好。
- 金利↑ or 安定予想 → 短期債を選好。

---

👉 デュレーションは「債券価格の金利感応度」を測る指標。  
👉 金利リスク管理に不可欠。


##2-5 債券のリスク

### 1. 金利リスク（Interest Rate Risk）
- 債券価格と市場金利は逆相関。
- 金利↑ → 債券価格↓
- 満期まで保有すれば影響はないが、途中売却時に損失リスク。

---

### 2. デフォルトリスク（Default Risk）
- 発行体（企業・政府）が利息や元本を支払えないリスク。
- 例：クーポン（年間$100）が支払われない可能性。

---

### 3. インフレリスク（Inflation Risk）
- インフレにより将来キャッシュフローの購買力が低下するリスク。
- 名目額を受け取っても実質価値は目減りする。

---

👉 債券投資では「金利リスク・デフォルトリスク・インフレリスク」の3つが特に重要。


##2-6 株式と債券の違い

### 株式（Stocks）
- 企業の所有権を表す。
- 株主は議決権を持つ。
- 配当（Dividends）が支払われることがある。
- 債券よりリスクが高い。

---

### 債券（Bonds）
- 政府・企業への「貸付」を表す負債証券。
- 満期時に元本と利子が返済される。
- 債券保有者は議決権を持たない。
- 配当はなく利息のみ。
- 債券価格は市場金利と逆相関。
- 株式より安全とされる。

---

### まとめ
- 株式 = 所有権（ハイリスク・ハイリターン）
- 債券 = 負債投資（ローリスク・安定収益）


##2-7 Python実装：ゼロクーポン債の価格計算

### 特徴
- クーポン（利払い）がない。
- 満期時に元本のみ支払われる。
- 価格は将来キャッシュフローの現在価値。

---

### 数式
$$
P = \frac{X}{(1+r)^N}
$$
- X = 元本（Principal）
- r = 市場金利（Market Interest Rate, 小数）
- N = 満期（年数）

---

### Python実装例
（下のコードセル参照）


```python
class ZeroCouponBond:
    def __init__(self, principal, maturity, interest_rate):
        self.principal = principal
        self.maturity = maturity
        self.interest_rate = interest_rate / 100  # % → 小数

    def present_value(self, X, n):
        return X / ((1 + self.interest_rate) ** n)

    def calculate_price(self):
        return self.present_value(self.principal, self.maturity)


if __name__ == "__main__":
    bond = ZeroCouponBond(1000, 2, 4)  # 元本1000, 満期2年, 金利4%
    price = bond.calculate_price()
    print(f"Price of the bond in dollars = {price:.2f}")
```

##2-8 Python実装：クーポン債の価格計算

### 特徴
- 毎年クーポン（利払い）あり。
- 満期時に元本を返済。
- 価格 = クーポン支払の現在価値 + 元本の現在価値。

---

### 数式
$$
P = \sum_{t=1}^N \frac{C}{(1+r)^t} + \frac{X}{(1+r)^N}
$$
- C = クーポン支払額（Principal × Coupon Rate）
- X = 元本（Principal）
- r = 市場金利（Market Interest Rate, 小数）
- N = 満期（年数）

---

### Python実装例
（下のコードセル参照）


```python
class CouponBond:
    def __init__(self, principal, rate, maturity, interest_rate):
        self.principal = principal
        self.rate = rate / 100          # クーポン利率
        self.maturity = maturity
        self.interest_rate = interest_rate / 100  # 市場金利

    def present_value(self, X, n):
        return X / ((1 + self.interest_rate) ** n)

    def calculate_price(self):
        price = 0
        coupon = self.principal * self.rate
        for t in range(1, self.maturity + 1):
            price += self.present_value(coupon, t)
        price += self.present_value(self.principal, self.maturity)
        return price


if __name__ == "__main__":
    bond = CouponBond(1000, 10, 3, 4)  # 元本1000, クーポン10%, 満期3年, 市場金利4%
    price = bond.calculate_price()
    print(f"Price of the bond = {price:.2f}")
```

##1-1お金の時間的価値
お金は現在と将来で、価値が違う可能性がある。
将来価格をFV、現在価格をPV、利率をr、期間をnとすると、

$$FV = PV × (1 + r)^n$$
$$PV = \frac{FV}{(1 + r)^n}$$
のようになる。このモデルは離散時間モデルといわれる。

次に、連続的な時間でのモデルを考える。
時刻tにおけるある資産の価値をx(t)とする。利率をrとすると、微小時間$dt$での増分は利率と元本に比例するので
$$\frac{dx}{dt} = r ⋅ x(t)$$
という微分方程式が成り立つ。
これを解くと、
$$FV = PV⋅e^{rt}$$
$$PV = FV⋅e^{-rt}$$
と求まる。
以上のことから、離散時間モデルでも連続時間モデルでも、お金は指数関数的に成長することが言える。

##1-2 Pythonによる現在価値と将来価値の計算

### 離散時間モデル
投資額を **X**、利率を **r**、年数を **n** とすると、

$$FV = X \times (1+r)^n$$
$$PV = \frac{X}{(1+r)^n}$$

### 連続時間モデル
時間を **t** とすると、

$$FV = X \cdot e^{rt}$$
$$PV = X \cdot e^{-rt}$$


```python
from math import exp

def future_discrete(X, r, n):
    return X * (1 + r) ** n

def present_discrete(X, r, n):
    return X * (1 + r) ** -n

def future_continuous(X, r, t):
    return X * exp(r * t)

def present_continuous(X, r, t):
    return X * exp(-r * t)


# パラメータ
X = 10000
r = 0.05
n = 10
t = 10

print("離散時間モデル")
print(f"今、{X} 円を年利 {r*100}% で {n} 年運用すると、{future_discrete(X, r, n):.2f} 円になる")
print(f"{n} 年後にもらえる {X} 円は、今の価値だと {present_discrete(X, r, n):.2f} 円である")

print("連続時間モデル")
print(f"今、{X} 円を年利 {r*100}% で {t} 年運用すると、{future_continuous(X, r, t):.2f} 円になる")
print(f"{t} 年後にもらえる {X} 円は、今の価値だと {present_continuous(X, r, t):.2f} 円である")
```

## 1-3 株式

### 株式とは
- 株式は**株式会社の所有権の一部**を表す金融商品。
- 株式の保有者（株主）は  
  - 配当を受け取る権利  
  - 株主総会での議決権    を持つ。
- 企業は自己資本による**資金調達**のために株式を発行する。

---

### 株式市場の仕組み
- **企業**：成長投資や事業拡大のために株式を発行（IPOや増資）。
- **証券取引所**：株式の売買が行われる市場（例：東京証券取引所）。
- **投資家**：株式を売買しリターンを得る主体。
- **証券会社（ブローカー）**：投資家の注文を取引所に取り次ぐ。
---

### 株価の変動要因
- 株価は**需要と供給**のバランスによって決まる。
- 買い注文が多ければ株価は上昇し、売り注文が多ければ下落する。
- 理論的には、短期の株価変動は  
  **ランダムウォーク**に近い性質を持つとされる。
  - これは「将来の価格を完全に予測できない」ことを意味する。
  - ただし、実務・研究では完全なランダムではない歪み（アノマリー）も研究されている。

---

### 株を持つメリット
1. **キャピタルゲイン（値上がり益）**
   - 株価が上昇した後に売却することで得られる利益。
   - 例：**トヨタ自動車**
     - 2012年頃：約3000円台  
     - 2023年頃：7000円前後  
     → 長期的に大きな値上がりを経験。

2. **配当金（Dividends）**
   - 企業の利益の一部を株主に還元。
   - 日本企業では年1回または年2回支払われることが多い。
   - 安定企業（例：NTT、三菱UFJなど）は配当重視の投資対象になる。

---

### 株のリスクとボラティリティ
- 株式は**価格変動リスクの高い資産**に分類される。
- **ボラティリティ（Volatility）**
  - リターンのばらつきを表す指標。
  - 主に **分散・標準偏差** で測定される。
  - 高ボラティリティ = 価格変動が大きく、リスクが高い。
- 4章で述べる：
  - **CAPM**（資本資産価格モデル）
  - **β値**による市場リスク評価


## 1-4 コモディティ（Commodities）

### コモディティとは
- 金、銀、原油、天然ガス、農産物などの**原材料・資源**。
- 一般に価格変動が大きく、**ボラティリティが高い資産クラス**。
- 株式と異なり、保管・輸送の問題があるため  
  **現物取引よりもデリバティブ（先物）取引が中心**となる。

---

### 価格変動の要因
- 基本は **需給（供給と需要）** によって価格が決まる。
- 需要増加や供給減少 → 価格上昇  
  供給増加や需要減少 → 価格下落
- 天候、地政学リスク、在庫水準、金融政策などの影響も受けやすい。
- 短期的な価格推移は**ランダムウォーク**に近い確率的性質を持つとされる。

---

### 先物契約（Futures）
- 将来の特定時点に、あらかじめ決めた価格で  
  商品を売買する契約。
- **価格変動のリスクをヘッジ**する目的で利用される。
  - 例：航空会社が燃料費高騰を避けるために原油先物を購入。
- 主な先物市場：
  - 大阪取引所（OSE）
  - 東京商品取引所（旧TOCOM）
  - ニューヨーク・マーカンタイル取引所（NYMEX）
  - シカゴ・マーカンタイル取引所（CME）
---

### 投資方法
1. **直接投資**
   - 先物市場やブローカーを通じて先物契約を取引。
   - レバレッジがかかるためリスクは高い。
2. **間接投資**
   - コモディティ価格に依存する企業の株式を購入。
   - 例：  
     - 金価格 ↔ 金鉱株（正の相関が見られることが多い）

---

### ポートフォリオにおける役割
- **インフレ対策**
  - インフレ局面では商品価格が上昇しやすい傾向がある。
- **分散投資効果**
  - 株式・債券と異なる値動きをするため、  
    ポートフォリオ全体のリスク低減に寄与。
- **戦略的利用**
  - コモディティ価格と株価の相関を利用した  
    ペアトレードやヘッジ戦略も用いられる。

---

👉 コモディティは**高リスク資産**である一方、**インフレ防衛・分散投資**の観点で重要な資産クラスである。


##1-5 通貨と外国為替市場（Forex Market）

### 為替レートとは
- ある通貨を別の通貨に交換する際の**交換比率**。
- 基軸通貨（Base Currency）・ 相手通貨（Quote Currency） の形で表される。
- 為替レートが上昇  
  → 基軸通貨高・相手通貨安  
- 為替レートが下落  
  → 基軸通貨安・相手通貨高
- 例：  
  -1ドル = 110円 → 1ドル = 120円 これは ドル高・円安

---

### 為替市場（Forex）
- **Forex（Foreign Exchange）**：外国為替市場。
- 各国通貨（USD, EUR, JPY, CHF など）を取引する市場。
- 世界最大の金融市場であり、**24時間取引・OTC（相対取引）市場**が中心。
- 個人投資家は **FXブローカー**を通じて市場に参加する。

---

### 為替レートに影響する要因
1. **金利（Interest Rates）**
   - 中央銀行が政策金利を通じて操作。
   - 一般に、金利上昇 → 通貨需要増加 → 通貨高。
2. **マネーサプライ（Money Supply）**
   - 通貨供給量の増加はインフレ圧力を生みやすい。
   - 通貨供給↑ → 通貨価値↓（通貨安）。
3. **金融・政治の安定性（Financial Stability）**
   - 政治・経済が安定している国の通貨は 「安全通貨」として選好されやすい。

---

### アービトラージ（Arbitrage）
- 複数通貨間の交換を利用し、理論上リスクなしで利益を得る取引。
- 数理モデル：
  - 通貨を **ノード（頂点）**
  - 為替レートを **エッジ（辺）** とする有向グラフで表現
- 数学的変換：
  - エッジ重み = `-log(為替レート)`
  - 掛け算構造を足し算に変換することで解析を容易にする
- **Bellman–Fordアルゴリズム**を用いて  
  - **負の閉路（Negative Cycle）** を検出できれば  
    → アービトラージ機会が存在することを意味する
- 実務では、取引コスト・スプレッド・約定遅延の影響により  
  アービトラージ機会は極めて短命であり、主に **高頻度取引（HFT）** の分野で活用される


---

👉 為替市場では、需給・金利・マネーサプライ・金融安定性が通貨価値を左右する。  
👉 アービトラージは理論的には存在するが、実際には高度なアルゴリズムと高速執行が必要となる。


## 1-6 ロングポジションとショートポジション

### ロングポジション（Long Position）
- 株式や債券などの資産を**購入して保有**するポジション。
- 価格が上昇すれば利益、下落すれば損失が発生する。
- 最大損失は投資額まで。
- リスクが限定されており、初心者にも理解しやすい取引形態。

---

### ショートポジション（Short Position）
- 資産価格の**下落から利益を得る**ことを目的としたポジション。
- 主な形態：
  1. **保有資産の売却**
     - すでに持っている資産を売却し、価格下落リスクを回避。
  2. **空売り（Short Selling）**
     - 株式を借りて売却し、その後価格が下がった時点で買い戻して返却する。
     - 売値と買値の差が利益となる。
- 株価が上昇すると損失が拡大し、 理論上、損失の上限は存在しない。
- 一般にロングポジションよりも**リスクが高い**。

---

### 市場環境とポジション
- **Bull Market（強気相場）**
  - 価格上昇が持続する局面。
  - ロングポジションが有利になりやすい。
- **Bear Market（弱気相場）**
  - 価格下落が持続する局面。
  - ショートポジションが有効な場合がある。
- 市場は長期的に、「ブル相場とベア相場を繰り返す」と考えられている。

---

👉 ロングは「価格上昇で利益」、ショートは「価格下落で利益」。  
👉 ただし、ショートは損失が無制限になり得るため、ロングよりもリスク管理が重要である。

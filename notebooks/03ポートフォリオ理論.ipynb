#3 現代ポートフォリオ理論

##3-1 統計的基礎：期待値・分散・共分散・相関

### 期待値（Expected Value / Mean）
- 平均的な結果を表す。
- 式：
  $$E[X] = \sum_i x_i p_i$$
- 例：
  - サイコロ：期待値 = 3.5
  - ルーレット：期待値 < 0（長期的には損をする）

---

### 分散（Variance）と標準偏差（Stdard Deviation）
- データの散らばりを表す。
- 式：
  $$Var(X) = E[(X-\mu)^2]$$
- ただし$\mu=E[X]$
- 標準偏差 = 分散の平方根。
- 式：
  $$\sigma =  \sqrt{Var(X)}$$
---

### 共分散（Covariance）
- 2変数の一緒の動きを測る。
- 式：
  $$Cov(X,Y) = E[(X-\mu_X)(Y-\mu_Y)]$$
- 問題点：次元依存・比較が困難。

---

### 相関（Correlation）
- 共分散を標準偏差で正規化。
- 式：
  $$\rho = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}$$
- 値域：[-1, +1]
  - +1 → 正の相関
  - -1 → 負の相関
  - 0 → 無相関
- ポートフォリオ最適化では「相関が小さい銘柄の組み合わせ」が有効。



##3-2 現代ポートフォリオ理論（Markowitz Model）

### 概要
- Harry Markowitz（1952年提唱, ノーベル賞受賞）。
- 複数資産の組合せによるポートフォリオ最適化モデル。
- リスクを抑え、リターンを最大化するように構築する。

---

### 基本的考え方
- 単一銘柄投資は高リスク。
- 複数資産を組み合わせることで**リスクを分散**（Diversification）。
- 相関が低い資産を組み合わせるとリスクがさらに低減。

---

### 前提条件
1. リターンは正規分布に従う。
2. 投資家はリスク回避的（高リターンなら高リスクを取る）。

---

### 効率的ポートフォリオ（Efficient Portfolio）
- 与えられたリスクで最大リターン。
- または、与えられたリターンで最小リスク。
- 投資家は自分のリスク許容度に応じて最適な組合せを選択可能。



##3-3 現代ポートフォリオ理論の数式定式化

### ポートフォリオの前提
- ショートポジション禁止（全てロング）。
- 投資比率（weights）は非負で合計1。
- 投資資産 は 株式。

- 例：
  - トヨタ, Sony, NTT, 任天堂の株を次の比率で持つ
$w = [0.2, 0.3, 0.25, 0.25] , Σ_iw = 1.0$
---

### 基本パラメータ
- $w_i$ = 資産iの投資比率
- $r_i$ = 資産iのリターン（過去データから算出）
- $\mu_i$ = 資産iの期待リターン（平均）

---

### リターンの定義
1. **通常リターン**
   $$
   r_t = \frac{S_{t+1} - S_t}{S_t}
   $$
   - 株価の変化率。

2. **対数リターン**
   $$
   r_t = \ln\left(\frac{S_{t+1}}{S_t}\right)
   $$
   - 数学的に便利、正規化や機械学習で広く利用。

---

### まとめ
- 投資比率、リターン、期待リターン がポートフォリオの数式の基礎。
- リターン計算では対数リターンを使う方が一般的。


##3-4 ポートフォリオの期待リターン

### 前提
- 現代ポートフォリオでは「平均リターン」と「分散」を用いて効率的ポートフォリオを求める
- 過去の平均リターンを将来の推定値とみなす

---

### 定義
- $w_i$ = 資産iの投資比率（weights）
- $r_i$ = 資産iのリターン
- $μ_i = E[r_i]$（資産iの期待リターン）

---

### 数式
ポートフォリオの期待リターン：
$$
\mu_{portfolio}=E[\sum_i w_i \, r_i] = \sum_i w_i \, μ_i=w^T μ
$$


---

### ポイント
- 期待リターンは「各資産の期待リターン × 投資比率」の合計
- 投資比率 w の合計は 1


##3-5 ポートフォリオのリスク

### 共分散
- 2つの資産のリターンの連動性を表す
- $Cov < 0$ → 逆相関（リスク低減に有効）
- $Cov > 0$ → 同方向に動く（分散効果が小さい）

### 共分散行列 Σ
- 対角要素：各銘柄の分散 $$\sigma_{i}^2 = E[(r_i-\mu_i)^2]$$
- 非対角要素：銘柄間の共分散 $$\sigma_{ij} = E[(r_i-\mu_i)(r_j-\mu_j)]$$
- 対称行列

### ポートフォリオ分散
$$
\sigma_{portfolio}^2 = w^T \Sigma w
$$

### 標準偏差（リスク）
$$
\sigma_{portfolio} = \sqrt{w^T \Sigma w}
$$

---

### ポイント
- 分散（リスク）は銘柄間の共分散も含めて計算される。
- 相関が低い資産を組み合わせることでリスクを下げられる。
- 現代ポートフォリオ理論の目的は「リスク最小でリターン最大」の効率的ポートフォリオを見つけること。


##3-6 効率的フロンティア

- 横軸：リスク（標準偏差・分散）
- 縦軸：リターン（期待収益率）
- 各点：異なるウェイトのポートフォリオ

### 効率的フロンティアとは
- 同じリスクでより高いリターンを得られるポートフォリオ
- 同じリターンでより低いリスクを持つポートフォリオ
- 最適なポートフォリオの集合が「フロンティア」を形成する

### 投資家の選択
- リスクを抑えるなら → 低リスク・低リターン側
- 高いリターンを狙うなら → 高リスク・高リターン側
- 自分のリスク許容度に応じてフロンティア上から選択

### まとめ
- リターンを得るには必ずリスクを取る必要がある
- 現代ポートフォリオ理論は「リスクとリターンのバランスを最適化する」ための理論


##3-7 Python実装：シャープレシオによる投資評価

### 特徴
- 投資のリスクとリターンを同時に評価できる指標。
- 期待リターンから無リスク利子率を引き、リスク（標準偏差）で割った値。
- 値が大きいほど投資効率が良い。

---

### 数式
$$S(w) = \frac{r_w - R_f}{\sigma(w)}$$

- w:投資比率
- $r_w$:ポートフォリオの期待リターン
- $R_f$:無リスク利子率
- $\sigma(w)$:ポートフォリオのリスク
---

### 解釈の目安
- $S(w) > 1$ : 良い投資
- $S(w) > 2$ : とても良い投資
- $S(w) > 3$ : 優れた投資（例：ウォーレン・バフェットの投資）

---

### Python実装例


```python
#Sharpe Ratio（シャープレシオ）を計算する関数。
def sharpe_ratio(returns, risk_free_rate):
  import numpy as np
  returns = np.array(returns)                   # ❶ returns を numpy 配列へ変換
  excess_returns = returns - risk_free_rate     # ❷ 超過リターン（= リターン − 無リスク利子率）ポートフォリオがリスクなし資産との差でどれだけ儲かったか
  mean_excess_return = np.mean(excess_returns)  # ❸ 超過リターンの平均
  risk = np.std(excess_returns, ddof=1)         # ❹ 超過リターンの標準偏差（リスク）    ddof=1 は標本標準偏差（実データから推定する際に使う）
  sharpe = mean_excess_return / risk            # ❺ シャープレシオ = 平均超過リターン / リスク  → リスク1単位あたりどれだけ稼いでいるか？
  return sharpe

# 実行部分
if __name__ == "__main__":
    returns = [0.05, 0.07, 0.02, 0.06, 0.04]     # 年間の実現リターン例
    rf = 0.01                                    # 無リスク利子率（1%）
    sr = sharpe_ratio(returns, rf)               # シャープレシオ計算
    print("Sharpe Ratio:", sr)
```

##3-8 Python実装：キャピタル・アロケーション・ライン（Capital Allocation Line, CAL）

### 特徴
- 投資家は「リスク資産（株式ポートフォリオ）」と「無リスク資産（国債やTビル）」を組み合わせて投資できる。
- 無リスク資産のリターンと、最大シャープレシオを持つポートフォリオを結ぶ直線を「キャピタル・アロケーション・ライン（CAL）」と呼ぶ。
- CAL上にあるポートフォリオが最適ポートフォリオとなる。

---

### 考え方
- リスクを取りたくない → 無リスク資産のみ（CALの左端）。
- 一部リスクを取る → 無リスク資産とリスク資産を組み合わせた点（CAL上の中間）。
- 全額リスク資産 → 最大シャープレシオを持つポートフォリオ（CALの接点）。
- さらにリスクを取りたい → 借入をしてリスク資産に追加投資（CALの右側延長）。

---

### 数式
CAL上の任意のポートフォリオの期待リターンは次の式で表される。

$$E[R_c] = R_f + ( (E[R_p] - R_f) / σ_p ) * σ_c$$

- $E[R_c]$ : CAL上のポートフォリオの期待リターン  
- $R_f$ : 無リスク利子率  
- $E[R_p]$ : リスク資産ポートフォリオの期待リターン  
- $σ_p$ : リスク資産ポートフォリオの標準偏差  
- $σ_c$ : CAL上のポートフォリオの標準偏差  

---

### Python実装例


```python
#CAL（資本配分線）に基づき、任意のポートフォリオの期待リターンを計算する関数
def capital_allocation_line(rf, rp, sigma_p, sigma_c):
    slope = (rp - rf) / sigma_p  # ❶ CAL の傾き（= リスク資産のシャープレシオ）無リスク資産とリスク資産の組み合わせで得られる効率性を示す
    return rf + slope * sigma_c  # ❷ CAL の式： 期待リターン = rf + slope * sigma_c    任意のリスク量 sigma_c に応じた期待リターンが計算できる

# 実行部分
if __name__ == "__main__":
    rf = 0.02                                      # 無リスク利子率（2%）
    rp = 0.08                                      #ポートフォリオの期待リターン（8%）
    sigma_p = 0.15                                 # リスク資産ポートフォリオのリスク（15%）

    # 投資家が選ぶリスク量（標準偏差）
    for sigma_c in [0.0, 0.05, 0.1, 0.15, 0.2]:    #計算したいポートフォリオの標準偏差
        rc = capital_allocation_line(rf, rp, sigma_p, sigma_c)
        print(f"σ={sigma_c:.2f}, 期待リターン={rc:.4f}")
```

##3-10 Python実装：リターンとポートフォリオ統計量の計算

### 特徴
- 株価データから対数リターン（log return）を計算する。
- 年間平均リターン、分散・共分散行列を算出する。
- 任意のポートフォリオ（重みベクトル）に対して、期待リターンとリスク（ボラティリティ）を計算する。

---

### 数式と定義
対数リターン（Log Return）:  
$$r_t = ln\frac{S_t}{S_{t-1}}$$

ポートフォリオ期待リターン:  
$$E[R_p] = (μ ・ w) × 252  $$
- $μ$: ポートフォリオの重みベクトル  
- 252 : 年間取引日数  

ポートフォリオ分散:  
$$Var(R_p) = w^T Σ w × 252  $$
- $Σ$ : リターンの共分散行列  
- $w^T$ : 重みベクトルの転置  

ポートフォリオ標準偏差（ボラティリティ）:  
$$σ_p = \sqrt{Var(R_p)}$$

---

### Python実装例


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

TRADING_DAYS = 252

def download_data(stocks, start, end):
    data = {}
    for stock in stocks:
        ticker = yf.Ticker(stock)
        data[stock] = ticker.history(start=start, end=end)["Close"]
    return pd.DataFrame(data)

def calculate_log_returns(data):
    log_returns = np.log(data / data.shift(1))
    return log_returns.dropna()

def show_statistics(returns):
    mean_returns = returns.mean() * TRADING_DAYS
    cov_matrix = returns.cov() * TRADING_DAYS
    print("Annual Mean Returns:\n", mean_returns)
    print("\nAnnual Covariance Matrix:\n", cov_matrix)
    return mean_returns, cov_matrix

def portfolio_performance(weights, mean_returns, cov_matrix):
    port_return = np.dot(mean_returns, weights)
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)
    return port_return, port_volatility


if __name__ == "__main__":
    stocks = ["7203.T", "6758.T", "9432.T", "7974.T"]

    name_map = {
        "7203.T": "Toyota",
        "6758.T": "Sony",
        "9432.T": "NTT",
        "7974.T": "Nintendo"
    }

    start = "2020-01-01"
    end = "2024-01-01"

    # ① データ取得
    data = download_data(stocks, start, end)

    # ② ★ここで名前変更（超重要）
    data = data.rename(columns=name_map)

    # ③ 以降の計算は「会社名」で進む
    log_returns = calculate_log_returns(data)
    mean_returns, cov_matrix = show_statistics(log_returns)

    weights = np.array([1 / len(data.columns)] * len(data.columns))

    port_return, port_volatility = portfolio_performance(
        weights, mean_returns, cov_matrix
    )

    print("\nPortfolio Expected Annual Return:", round(port_return, 4))
    print("Portfolio Volatility (Std Dev):", round(port_volatility, 4))
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

TRADING_DAYS = 252  # 年間取引日数

def download_data(stocks, start, end):
    data = {}
    for stock in stocks:
        ticker = yf.Ticker(stock)
        data[stock] = ticker.history(start=start, end=end)["Close"]  # 終値のみ取得
    return pd.DataFrame(data)

def calculate_log_returns(data):
    log_returns = np.log(data / data.shift(1))  # 対数リターン
    return log_returns.iloc[1:]                 # 1行目はNaNなので除外

def show_statistics(returns):
    mean_returns = returns.mean() * TRADING_DAYS        # 年率平均リターン
    cov_matrix = returns.cov() * TRADING_DAYS           # 年率共分散
    print("Annual Mean Returns:\n", mean_returns)
    print("\nAnnual Covariance Matrix:\n", cov_matrix)
    return mean_returns, cov_matrix

def portfolio_performance(weights, mean_returns, cov_matrix):
    port_return = np.dot(mean_returns, weights)                 # 年率期待リターン
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)                    # 標準偏差（リスク）
    return port_return, port_volatility

if __name__ == "__main__":
    stocks = ["AAPL", "WMT", "AMZN", "DB", "GE", "TSLA"]
    start = "2012-01-01"
    end = "2017-01-01"

    data = download_data(stocks, start, end)
    log_returns = calculate_log_returns(data)

    mean_returns, cov_matrix = show_statistics(log_returns)

    weights = np.array([1/len(stocks)] * len(stocks))  # 全銘柄均等
    port_return, port_volatility = portfolio_performance(weights, mean_returns, cov_matrix)

    print("\nPortfolio Expected Annual Return:", round(port_return, 4))
    print("Portfolio Volatility (Std Dev):", round(port_volatility, 4))
```

##3-11 Python実装：ランダムポートフォリオ生成と効率的フロンティア

### 特徴
- ランダムに重み（weights）を割り当てた多数のポートフォリオを生成する。
- 各ポートフォリオについて期待リターンとリスク（ボラティリティ）を計算。
- 結果を散布図で表示し、効率的フロンティア（Efficient Frontier）を可視化。
- カラーバーでシャープレシオを示す。

---

### 数式
ポートフォリオ期待リターン:  
$$E[R_p] = (μ ・ w) × 252  $$

ポートフォリオ分散:  
$$Var(R_p) = w^T Σ w × 252  $$

ポートフォリオ標準偏差（ボラティリティ）:  
$$σ_p = sqrt(Var(R_p))  $$

シャープレシオ（カラーの基準）:  
$$Sharpe Ratio = E[R_p] / σ_p  $$

---

### Python実装例


```python
import numpy as np
import matplotlib.pyplot as plt

TRADING_DAYS = 252  # 年間取引日数（mean_returns, cov_matrix は年率前提）

def generate_portfolios(mean_returns, cov_matrix, num_portfolios=10000):
    """ランダムにポートフォリオを生成し、期待リターン・リスク・重みを返す"""
    results = {"returns": [], "risks": [], "weights": []}
    num_stocks = len(mean_returns)

    for _ in range(num_portfolios):
        weights = np.random.random(num_stocks)   # ランダム重み
        weights /= np.sum(weights)               # 合計が1になるよう正規化

        port_return = np.dot(mean_returns, weights)                 # ポートフォリオ期待リターン
        port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))  # 分散
        port_volatility = np.sqrt(port_variance)                    # 標準偏差（ボラティリティ）

        results["returns"].append(port_return)
        results["risks"].append(port_volatility)
        results["weights"].append(weights)

    return (np.array(results["returns"]),
            np.array(results["risks"]),
            np.array(results["weights"]))

def show_portfolios(returns, risks):
    """ランダムポートフォリオを散布図で表示（カラーはシャープレシオ）"""
    plt.figure(figsize=(10, 6))
    plt.scatter(risks, returns, c=returns/risks, marker='o')  # c=Sharpe Ratio
    plt.xlabel("Expected Volatility")
    plt.ylabel("Expected Return")
    plt.colorbar(label="Sharpe Ratio")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # ダミーデータ例（実際は前の講義で計算した年率 mean_returns, cov_matrix を使う）
    mean_returns = np.array([0.12, 0.08, 0.15, 0.03])  # 各資産の年率期待リターン
    cov_matrix = np.array([
        [0.04, 0.01, 0.02, 0.0],
        [0.01, 0.03, 0.01, 0.0],
        [0.02, 0.01, 0.05, 0.0],
        [0.0,  0.0,  0.0,  0.02]
    ])  # 年率共分散行列

    port_returns, port_risks, port_weights = generate_portfolios(mean_returns, cov_matrix, 5000)
    show_portfolios(port_returns, port_risks)
```

##3-12 Python実装：最適ポートフォリオとシャープレシオ最大化

### 特徴
- SciPy の最適化アルゴリズムを用いて、**シャープレシオが最大となるポートフォリオ**を求める。
- 制約条件：
  - 各資産の重みは 0〜1 の範囲。
  - 重みの合計は 1。
- 最適化は「最小化」をベースにしているため、`-Sharpe Ratio` を最小化することでシャープレシオを最大化する。

---

### 数式
ポートフォリオ期待リターン:  
$$E[R_p] = (μ ・ w) × 252  $$

ポートフォリオ標準偏差（ボラティリティ）:  
$$σ_p = sqrt( w^T Σ w × 252 )  $$

シャープレシオ:  
$$Sharpe Ratio = E[R_p] / σ_p  $$

---

### Python実装例


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.optimize import minimize

TRADING_DAYS = 252  # 年間取引日数（mean_returns, cov_matrix は年率前提）

def statistics(weights, mean_returns, cov_matrix):
    """与えられた重みでポートフォリオ指標を計算"""
    port_return = np.dot(mean_returns, weights)                    # 年率期待リターン
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights)) # 年率分散
    port_volatility = np.sqrt(port_variance)                       # 年率ボラティリティ
    sharpe_ratio = port_return / port_volatility                   # シャープレシオ（rf=0仮定）
    return np.array([port_return, port_volatility, sharpe_ratio])

def min_func_sharpe(weights, mean_returns, cov_matrix):
    """シャープレシオを最大化したいので、そのマイナスを最小化"""
    return -statistics(weights, mean_returns, cov_matrix)[2]

def optimize_portfolio(mean_returns, cov_matrix):
    """シャープレシオ最大ポートフォリオを最適化で求める"""
    num_stocks = len(mean_returns)
    args = (mean_returns, cov_matrix)

    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})  # 重みの合計=1
    bounds = tuple((0, 1) for _ in range(num_stocks))              # 各資産 0〜1
    init_guess = num_stocks * [1. / num_stocks]                    # 初期値：均等重み

    result = minimize(min_func_sharpe,
                      init_guess,
                      args=args,
                      method="SLSQP",
                      bounds=bounds,
                      constraints=constraints)
    return result

def print_optimal_portfolio(opt_result, mean_returns, cov_matrix):
    """最適ポートフォリオの結果を表示"""
    opt_weights = opt_result["x"].round(3)
    ret, vol, sharpe = statistics(opt_weights, mean_returns, cov_matrix)
    print("Optimal Weights:", opt_weights)
    print("Expected Return:", round(ret, 4))
    print("Volatility:", round(vol, 4))
    print("Sharpe Ratio:", round(sharpe, 4))

if __name__ == "__main__":
    # ダミーデータ例（実際は前のステップで計算した値を使う）
    mean_returns = np.array([0.12, 0.08, 0.15, 0.03])  # 各資産の年率期待リターン
    cov_matrix = np.array([
        [0.04, 0.01, 0.02, 0.0],
        [0.01, 0.03, 0.01, 0.0],
        [0.02, 0.01, 0.05, 0.0],
        [0.0,  0.0,  0.0,  0.02]
    ])  # 年率共分散行列

    opt_result = optimize_portfolio(mean_returns, cov_matrix)
    print_optimal_portfolio(opt_result, mean_returns, cov_matrix)
```

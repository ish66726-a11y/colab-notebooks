#4 CAPM(Capital Asset Pricing Model)資本資産評価モデル

#4-1 CAPM入門：リスクの種類とβの重要性

### 背景
- これまでモダンポートフォリオ理論（Markowitzモデル）を実装し、最適ポートフォリオを構築した
- 複数の株式を組み合わせることでリスクを低減できることを学んだ
- しかし、すべてのリスクを排除することはできない

---

### リスクの2種類
1. **非体系的リスク（Unsystematic Risk / Specific Risk）**
   - 個別銘柄に固有のリスク（企業業績、経営問題など）
   - 分散投資によりリスクを軽減・消去可能
   - Markowitzモデルで扱っているのはこのリスク
   - 例：Microsoft株に全額投資 → 企業固有のリスクを強く受ける

2. **体系的リスク（Systematic Risk / Market Risk）**
   - 分散投資しても消せない市場全体のリスク
   - 要因：金利変動、不況、戦争など
   - 例：金利上昇により債券価格が下落する「金利リスク」
   - 将来予測が困難で、投資家が必ず直面するリスク

---

### CAPMとβパラメータ
- 資本資産評価モデル（CAPM）は「体系的リスク」を測定するモデル
- **β（ベータ）**：市場全体の動きに対する感応度を表す指標
  - β > 1 : 市場より価格変動が大きい（高リスク・高リターン）
  - β < 1 : 市場より価格変動が小さい（低リスク・低リターン）
  - β = 1 : 市場と同じ動き
- 投資家にとって重要なのは、分散で消せない「体系的リスク＝市場リスク」




#4-2 CAPMの基本式とβパラメータ

### CAPMの定義
- ウィリアム・シャープ（1960年代初頭）によって提唱
- 「期待収益率」と「市場プレミアム」の間に **線形関係** を定義するモデル

---

### 基本式
$$E[r_a] = r_f + β_a ( E[r_m] - r_f )$$

- $E[r_a]$ : 投資対象（株式やポートフォリオ）の期待収益率
- $r_f $: 無リスク利子率（国債・Tビル・銀行預金など）
- $E[r_m]$ : 市場全体（例：S&P500,TOPIX）の期待収益率
- $β_a $: 投資対象のリスク感応度（市場の揺れ1単位あたり、その資産が何単位揺れるか）

---

### ポイント
- **市場プレミアム** : $(E[r_m] - r_f) $
  - 投資家がリスクを取ることで“追加でもらえる期待リターン”は、全部市場プレミアムから配られている
- **線形関係** : 投資対象の期待リターンは、市場リスクプレミアムとβで決まる
- リスクを取らない（β=0）→ 無リスク利子率のみ
- リスクを取る（β>0）→ 市場リターンに応じて期待収益が増える

---

### βパラメータの計算
$$β_a = \frac{Cov(r_a, r_m)}{Var(r_m)}$$

- $Cov(r_a, r_m)$ : 投資対象のリターンと市場リターンの共分散
- $Var(r_m) $: 市場リターンの分散
- 解釈:
  | 対象       | β  | 意味      |
| -------- | -- | ------- |
| 市場そのもの   | 1  | 基準      |
| 市場より激しい株 | >1 | レバレッジ的  |
| 市場に鈍い株   | <1 | ディフェンシブ |
| 市場と無関係   | 0  | 独立      |


---

### まとめ
- 非体系的リスク（個別リスク）は分散投資で消せる
- 体系的リスク（市場リスク）は消せず、βで測定される
- CAPMにおいて、βは **株式リスクの唯一の有効な尺度**


##4-3 βパラメータの解釈とポートフォリオβの計算

### βの意味
- **βは市場に対する相対的リスク（ボラティリティ）を表す指標**
- 投資対象（株式やポートフォリオ）のリターンが、市場全体（例：S&P500,TOPIX）のリターンとどの程度連動するかを測る

---

### βの値と解釈
| βの値       | 市場との関係 | 市場が1%動いたとき | CAPM上の期待リターン                   | 解釈・特徴     |
| --------- | ------ | ---------- | ------------------------------ | --------- |
| β < 0     | 市場と逆方向 | 逆に動く       | 市場より低い(負もあり得る) | ヘッジ的資産、特殊 |
| β = 0     | 無相関    | ほぼ動かない     | $r_f$                         | 市場リスクなし   |
| 0 < β < 1 | 市場より鈍感 | 1%未満       | 市場より低い                         | ディフェンシブ   |
| β = 1     | 市場と同じ  | 約1%        | 市場平均                           | 市場ポートフォリオ |
| β > 1     | 市場より敏感 | 1%以上       | 市場より高い                         | ハイリスク     |
| β ≫ 1     | 非常に敏感  | 大きく変動      | 非常に高い                          | レバレッジ的    |

---

### 具体例
- **β = 0.5**
  - 市場が +10% → 投資は +5%
  - 市場が -2% → 投資は -1%
- **β = 1.5**
  - 市場が +10% → 投資は +15%
  - 市場が -2% → 投資は -3%

---

### 数式まとめ
単一株式のβ:

$$
\beta_a = \frac{\text{Cov}(r_a, r_m)}{\text{Var}(r_m)}
$$

ポートフォリオのβ:

$$
\beta_p = w_1 \beta_1 + w_2 \beta_2 + \dots + w_n \beta_n
$$

- $r_a$ : 投資対象のリターン  
- $r_m$ : 市場リターン（例：S&P500,TOPIX）  
- $w_a$ : ポートフォリオにおける各資産の比率（合計 = 1）  
- $\beta_a$ : 各資産のβ  

---

### 例：ポートフォリオβの計算
- 構成銘柄と比率:
  - トヨタ : 20%
  - Sony : 30%
  - NTT : 25%
  - 任天堂 : 25%
- 各株式のβを計算し、上式に代入して加重平均を取ることでポートフォリオ全体のβを算出

---

### まとめ
- βは市場リスク（体系的リスク）の唯一の有効な尺度
- βの値によって「市場よりリスクを取っているのか」「抑えているのか」が判断できる
- 複数銘柄を組み合わせたポートフォリオのβは、**各銘柄のβの加重平均**で求められる


#4-4 線形回帰（Linear Regression）の基礎

### 定義
- 線形回帰は、**目的変数（Y）** と **説明変数（X）** の関係を線形でモデル化する手法。
- シンプルな例：家の価格を家のサイズから予測する。

---

### 線形回帰の種類
1. **単回帰（Simple Linear Regression）**
   - 説明変数が1つ（例：家のサイズ → 家の価格）
2. **重回帰（Multiple Linear Regression）**
   - 説明変数が複数（例：家のサイズ、立地、部屋数 → 家の価格）

---

### モデル式
$$
H(x) = b_0 + b_1 x
$$

- $x$：説明変数（例：家のサイズ）  
- $H(x)$：モデルが予測する目的変数（例：価格）  
- $b_0, b_1$：学習によって推定されるパラメータ  

---

### 誤差と損失関数
各データ点の誤差:
$$
\epsilon_i = y_i - H(x_i)
$$

平均二乗誤差（MSE）:
$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - H(x_i))^2
$$

- 誤差が小さいほどモデル精度が高い。
- 損失関数を最小化することで最適な $b_0, b_1$を求める。

---

### 決定係数 $R^2$
回帰モデルの当てはまりを測る指標:
$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

- $SS_{res} = \sum (y_i - H(x_i))^2\$（残差平方和）  

- $SS_{tot} = \sum (y_i - \bar{y})^2\$（総平方和）  

- $R^2$ が1に近いほど強い線形関係を示す。
  - $R^2=0.8$→ 「Yの分散の80%をXで説明できている」

---

### まとめ
- 線形回帰はシンプルだが、CAPMにおける β推定 に応用可能。  
- 誤差を最小化することでモデルを最適化し、予測精度を高める。  
- $R^2$ を使えば「どの程度線形関係を説明できているか」がわかる。  


#4-5 CAPMと線形回帰の関係

### CAPMにおけるβの計算方法
1. **共分散と分散を使う方法**
$$
\beta_a = \frac{\text{Cov}(r_a, r_m)}{\text{Var}(r_m)}
$$

2. **線形回帰を使う方法**  
リスクフリーレート $r_f$ を引いた形に変形すると：
$$
E[r_a] - r_f = \alpha + \beta_a( E[r_m] - r_f )
$$

- $E[r_a]$：投資対象の期待収益率  
- $E[r_m]$：市場の期待収益率（例：S&P500）  
- $r_f$：無リスク利子率  
- $\alpha$：切片（実際の収益とCAPM推定値の差）  
- $\beta$：傾き（市場リスクに対する感応度）

---

### 線形回帰との対応
- 横軸（X）：市場超過リターン $(r_m - r_f)$
- 縦軸（Y）：投資対象超過リターン $(r_a - r_f)$  
- 回帰式：
$$
Y = \alpha + \beta X
$$
  - **傾き（β）** → システマティックリスク  
  - **切片（α）** → 実際のリターンと期待リターンの差

---

### α（アルファ）の意味
- CAPM理論式では $\alpha = 0$。  
- 実際の投資では：
  - **α > 0** → 実際のリターンがCAPM予測を上回る（優秀な投資）  
  - **α < 0** → 実際のリターンがCAPM予測を下回る（期待外れの投資）  

例：  
- CAPM推定リターン = 15%  
- 実際のリターン = 20%  
- → α = +5%  

---

### まとめ
- CAPMのβは **線形回帰の傾き** として推定可能。  
- βは市場リスク（体系的リスク）を測る唯一の尺度。  
- αはモデル予測との差を示し、投資の優劣を評価する指標。

##4-6 CAPM実装準備：データ取得とクラス設計

### 使用ライブラリ
- numpy
- pandas
- yfinance（Yahoo! Financeから株価データ取得）
- matplotlib

---

### 対象データ
- **市場インデックス**：S&P500（米国を代表する500社の株価指数 → 市場全体の代理）
- **銘柄例**：IBM
- **期間**：任意の開始日（start_date）と終了日（end_date）

---

### 実装ステップ
1. **CAPMクラスを作成**
   - 引数：`stocks`（IBMとS&P500）、`start_date`、`end_date`
   - 属性：株式リスト、期間、データ格納用変数

2. **データ取得メソッド**
   - `yfinance.download()` を利用
   - 調整後終値（Adjusted Close）を使用

3. **戻り値**
   - Pandas DataFrame に変換して管理

---

### Python実装例


```python
import numpy as np
import pandas as pd
import yfinance as yf

class CAPM:
    def __init__(self, stocks, start_date, end_date):
        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date
        self.data = None

    def download_data(self):
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,
            progress=False
        )

        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]
        else:
            close = df[["Close"]]
            close.columns = self.stocks

        self.data = close.dropna(how="any")
        return self.data


if __name__ == "__main__":
    # ✅ TOPIXそのもの(^TOPX / 998405.T)は yfinance で落ちやすいので、
    # ✅ TOPIX連動ETFを「市場」扱いにする（おすすめ）
    stocks = ["7203.T", "1306.T"]  # トヨタ（東証）, TOPIX連動ETF（市場プロキシ）

    capm = CAPM(stocks, "2015-01-01", "2020-01-01")
    df = capm.download_data()
    print(df.head())
```

##4-7 CAPM実装：βの算出（共分散アプローチ）

### ステップ概要
1. データ取得（IBM株とS&P500）
2. 日次データを月次データに変換（resample → 'M'）
3. 対数月次リターンを計算
4. 共分散行列を計算
5. βを算出
   $$
   \beta = \frac{\text{Cov}(R_{stock}, R_{market})}{\text{Var}(R_{market})}
   $$

---

### なぜ月次リターンを使うのか？
- **日次リターン**：短期的・戦術的分析に有効（休日や短期イベントを考慮）
- **月次リターン**：分布が正規分布に近づき、CAPMなど長期モデルに適合しやすい

---

### Python実装例



```python
import numpy as np
import pandas as pd
import yfinance as yf

class CAPM:
    def __init__(self, stocks, start_date, end_date):
        """
        stocks: [asset_ticker, market_ticker]
          例: ["7203.T", "1306.T"]
        """
        if len(stocks) != 2:
            raise ValueError("stocks は [個別銘柄, 市場] の2つで指定してください。")

        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date
        self.data = self.download_data()

    def download_data(self):
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,
            progress=False
        )

        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]
        else:
            close = df[["Close"]]
            close.columns = self.stocks

        # 2列そろっている日だけ残す
        return close.dropna(how="any")

    def calculate_beta(self, freq="M"):
        sampled = self.data.resample(freq).last()
        log_returns = np.log(sampled / sampled.shift(1)).dropna()

        asset_ret = log_returns[self.stocks[0]]
        market_ret = log_returns[self.stocks[1]]

        cov = np.cov(asset_ret, market_ret, ddof=1)
        beta = cov[0, 1] / cov[1, 1]
        return float(beta)


if __name__ == "__main__":
    # ✅ トヨタ × TOPIX（ETFを市場として使用）
    stocks = ["7203.T", "1306.T"]  # トヨタ, TOPIX連動ETF

    capm = CAPM(stocks, "2010-01-01", "2017-01-01")
    beta_value = capm.calculate_beta(freq="M")

    print("Asset :", stocks[0])
    print("Market:", stocks[1], "(TOPIX proxy)")
    print("Beta  :", round(beta_value, 3))
```

##4-8 CAPM実装：線形回帰によるβ・αの推定

### ステップ概要
1. 月次リターンを算出（株式リターン・市場リターン）
2. **線形回帰**で株式超過リターンと市場超過リターンの関係を推定  
   $$
   R_i - R_f = \alpha + \beta (R_m - R_f)
   $$
3. 推定したβとαから期待収益率を計算  
   $$
   E[R_i] = R_f + \beta (E[R_m] - R_f)
   $$
4. 回帰直線をプロットして可視化

---

### Python実装例


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

class CAPM:
    def __init__(self, stocks, start_date, end_date):
        """
        stocks = [個別銘柄, 市場（インデックス or ETF）]
          例: ["7203.T", "1306.T"]  # トヨタ × TOPIX連動ETF（市場プロキシ）
        """
        if len(stocks) != 2:
            raise ValueError("stocks は [個別銘柄, 市場] の2つで指定してください。")

        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date
        self.data = self.download_data()

    def download_data(self):
        """複数銘柄を一括ダウンロードして調整後終値相当（auto_adjust=True の Close）を取得"""
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,
            progress=False
        )

        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]
        else:
            close = df[["Close"]]
            close.columns = self.stocks

        # 2列そろっている日だけ残す（欠損で回帰が壊れるのを防ぐ）
        close = close.dropna(how="any")

        if close.empty:
            raise ValueError(
                "価格データが取得できませんでした。ティッカーが正しいか、期間を変えて再試行してください。"
            )

        return close

    def calculate_returns(self, freq="M"):
        """
        対数リターンを計算
        freq:
          "M" = 月末
          "W" = 週末
          "D" = 日次
        """
        sampled = self.data.resample(freq).last()
        log_returns = np.log(sampled / sampled.shift(1))
        return log_returns.dropna()

    def regression(self, freq="M"):
        """市場リターンと株式リターンで単回帰（α, β 推定）"""
        returns = self.calculate_returns(freq=freq)

        stock_returns = returns[self.stocks[0]]
        market_returns = returns[self.stocks[1]]

        # y = alpha + beta * x（x: 市場, y: 個別）
        beta, alpha = np.polyfit(market_returns, stock_returns, 1)
        return alpha, beta, stock_returns, market_returns

    def expected_return(self, rf=0.005, months_in_year=12, freq="M"):
        """
        CAPM による期待収益率（年率）を計算
        E[Ri] = Rf + β(E[Rm] - Rf)
        rf は年率の無リスク利子率（日本向けにデフォルト0.5%）
        """
        alpha, beta, stock_returns, market_returns = self.regression(freq=freq)

        # 市場の平均リターンを年率換算（freq="M"想定 → ×12）
        mean_market_return_annual = market_returns.mean() * months_in_year

        exp_return = rf + beta * (mean_market_return_annual - rf)
        return exp_return, alpha, beta

    def plot_regression(self, freq="M"):
        """市場リターン vs 株式リターンの散布図と回帰直線"""
        alpha, beta, stock_returns, market_returns = self.regression(freq=freq)

        plt.figure(figsize=(10, 6))
        plt.scatter(market_returns, stock_returns, label="Returns")
        plt.plot(
            market_returns,
            alpha + beta * market_returns,
            color="red",
            label="CAPM Regression Line"
        )
        plt.title(f"CAPM Regression: {self.stocks[0]} vs {self.stocks[1]} ({freq})")
        plt.xlabel("Market Returns")
        plt.ylabel("Stock Returns")
        plt.legend()
        plt.grid(True)
        plt.show()


if __name__ == "__main__":
    # ✅ 日本株：トヨタ（東証） × TOPIX（yfinanceで指数が取れないのでETFで代用）
    # 1306.T = NEXT FUNDS TOPIX連動型上場投信（TOPIXプロキシ）
    stocks = ["7203.T", "1306.T"]

    capm = CAPM(stocks, "2010-01-01", "2017-01-01")

    exp_return, alpha, beta = capm.expected_return(rf=0.005, freq="M")

    print(f"Asset : {stocks[0]}")
    print(f"Market: {stocks[1]} (TOPIX proxy)")
    print(f"Alpha : {alpha:.4f}, Beta: {beta:.4f}")
    print(f"Expected Annual Return (CAPM): {exp_return:.2%}")

    capm.plot_regression(freq="M")
```

##4-9 リターンの分布と正規分布仮定

### 目的
- 株価リターンが **正規分布に従う** と仮定してよいかを検証する。  
- ヒストグラムを描画し、正規分布との比較を行う。  

---

### 手順
1. IBM株価データ（2010〜2020年）の調整後終値を取得  
2. 対数日次リターンを計算  
   $$
   r_t = \ln\left(\frac{P_t}{P_{t-1}}\right)
   $$
3. 欠損値を削除  
4. リターンのヒストグラムを描画（例：700ビン）  
5. 平均・分散・標準偏差を算出  
6. 同じパラメータで正規分布曲線を重ね描画

### Python実装例


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.stats import norm

# ===== 設定 =====
asset_ticker = "7203.T"   # トヨタ
market_ticker = "1306.T"  # TOPIX連動ETF（TOPIXプロキシ）

start_date = "2010-01-01"
end_date = "2020-01-01"

# ===== 価格データ取得（調整後終値）=====
prices = yf.download(
    [asset_ticker, market_ticker],
    start=start_date,
    end=end_date,
    auto_adjust=True,
    progress=False
)["Close"]

# 欠損除去（両方そろっている日だけ）
prices = prices.dropna(how="any")

# ===== 日次対数リターン =====
log_returns = np.log(prices / prices.shift(1)).dropna()

asset_ret = log_returns[asset_ticker]
market_ret = log_returns[market_ticker]

# ===== 平均・標準偏差 =====
mu_asset, sigma_asset = asset_ret.mean(), asset_ret.std(ddof=1)
mu_market, sigma_market = market_ret.mean(), market_ret.std(ddof=1)

# ===== ヒストグラム =====
plt.figure(figsize=(10, 6))

plt.hist(
    asset_ret,
    bins=700,
    density=True,
    alpha=0.5,
    label="Toyota (7203.T)"
)

plt.hist(
    market_ret,
    bins=700,
    density=True,
    alpha=0.5,
    label="TOPIX (1306.T)"
)

# ===== 正規分布カーブ =====
x_asset = np.linspace(asset_ret.min(), asset_ret.max(), 1000)
x_market = np.linspace(market_ret.min(), market_ret.max(), 1000)

plt.plot(x_asset, norm.pdf(x_asset, mu_asset, sigma_asset),
         linewidth=2, label="Toyota Normal")

plt.plot(x_market, norm.pdf(x_market, mu_market, sigma_market),
         linewidth=2, label="TOPIX Normal")

plt.title("Daily Log-Return Distribution: Toyota vs TOPIX")
plt.xlabel("Log Returns")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# ===== 数値出力 =====
print("Toyota (7203.T)")
print(f"  mu    : {mu_asset:.6f}")
print(f"  sigma : {sigma_asset:.6f}")
print(f"  mu ± 3σ: [{mu_asset - 3*sigma_asset:.6f}, {mu_asset + 3*sigma_asset:.6f}]")

print("\nTOPIX (1306.T)")
print(f"  mu    : {mu_market:.6f}")
print(f"  sigma : {sigma_market:.6f}")
print(f"  mu ± 3σ: [{mu_market - 3*sigma_market:.6f}, {mu_market + 3*sigma_market:.6f}]")
```

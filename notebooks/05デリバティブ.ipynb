#5 デリバティブ

#5-1 デリバティブ入門：基礎と種類

### デリバティブとは？
- **定義**：価格が **原資産（Underlying Asset）** に依存する金融商品  
- **原資産の例**：株式、債券、通貨、金利、商品（原油など）
- **価格決定要因**：原資産の価格変動（確率的 → 確率微分方程式が必要）

---

### デリバティブの目的
1. **リスクヘッジ**
    
   - 将来の価格をあらかじめ固定 → 経営の安定化
   - 保険のように「リスクを回避」
   - 例：航空会社は燃料費（原油価格）の変動リスクを軽減するためにデリバティブを利用
2. **投機(Speculation)**
   - 将来の価格変動を予測して利益を狙う
   - 例：原油価格が上昇すると予想 → 先に契約して将来安く買い、高く売ることで利益獲得

---

### デリバティブの代表的な種類
1. **フォワード契約 (Forward)**
   - 2者間で将来、特定の価格で資産を売買する契約  
   - 規制なし、非標準化 → 取引所に上場されない  
   - 例：航空会社と石油会社が結ぶ長期燃料契約  

2. **先物契約 (Futures)**
   - フォワードと同様だが、**規制・標準化** されており、取引所で取引可能  

3. **スワップ (Swaps)**
   - 将来のキャッシュフローを交換する契約  
   - 例：金利スワップ（固定金利と変動金利の交換）  

4. **オプション (Options)**
   - **「権利」** だが **「義務」** ではない  
   - 特定の価格で将来、原資産を売買できる権利  
   - フォワードとの違い：オプションは行使するかどうかを選べる  

---

### まとめ
- デリバティブは原資産価格に依存する金融商品  
- 主な用途は「リスク回避（ヘッジ）」と「利益追求（投機）」  
- 種類：フォワード、先物、スワップ、オプション  

#5-2 フォワード契約と先物契約

### 共通点
- 双方とも「将来の特定時点に、あらかじめ決めた価格で資産を売買する契約」
- **目的**：将来の価格変動リスクを回避（ヘッジ）

---

### フォワード契約 (Forwards)
- **特徴**
  - 当事者同士の **私的な契約**
  - **非標準化・非規制** → 取引所では取引されない
  - 第三者の保証がない → **相手方が契約を履行しないリスク（信用リスク）** が存在

---

### 先物契約 (Futures)
- **特徴**
  - 内容はフォワード契約とほぼ同じ
  - ただし **標準化・規制** され、**取引所で売買可能**
  - 清算機関（取引所）が間に入り、契約履行を保証 → 信用リスクが低い

---

### 具体例

- **状況**
  - Aさん：トヨタ自動車の株を **5,000株保有**
  - 現在の株価：**1株 2,500円**
  - Aさんは将来の株価下落を心配している
  - Bさんは将来の株価上昇を予想し、買いのポジションを取りたい

- **契約内容**
  - **1年後に**
  - BさんがAさんから
  - **1株あたり 2,500円で**
  - **トヨタ株 5,000株を購入する先物契約**を締結


- **結果シナリオ**

  - **株価が下落した場合**
    - 市場価格：2,000円
    - Aさん：市場では損だが、契約により **2,500円で売却可能**
      → **価格下落リスクを回避（ヘッジ成功）**
    - Bさん：割高で買うため損失

  - **株価が上昇した場合**
    - 市場価格：3,000円
    - Aさん：上昇分の利益は得られない
    - Bさん：**2,500円で買って市場で売却**
     → **1株あたり500円の利益**
---

### まとめ
- フォワードと先物の基本的な仕組みは同じ  
- 違いは「規制・標準化」「取引所の有無」「信用リスクの有無」  
- 先物はヘッジにも投機にも広く利用される強力な金融商品  


#5-3 スワップ（Swaps）と金利スワップ

### スワップとは
- 2者間で **将来のキャッシュフローを交換する金融派生商品**
- 最初のスワップ取引は1980年代初頭、IBMと世界銀行の間で誕生
- 主な種類：
  - **金利スワップ（Interest Rate Swap）** ← 最も基本的で一般的
  - **通貨スワップ（Currency Swap）**：異なる通貨建ての元本＋利息を交換
  - **クレジット・デフォルト・スワップ（CDS）**：債務不履行時に保証（2008年金融危機で重要な役割）

---

### 金利スワップの目的
- 市場金利は常に変動し、借入コストに影響する
- **固定金利**：返済額が一定  
- **変動金利（例：LIBOR連動）**：市場金利の上下で返済額が変動 → リスク大  
- 金利スワップにより、当事者同士のニーズを調整して **支払条件を有利に変換**

---

### 具体例
- **会社A**：信用度が高い  
  - 固定金利で6%で借入可能  
  - または変動金利（= LIBOR）で借入可能  
  - → 実際は **変動金利を希望**
- **会社B**：信用度が低い  
  - 固定金利で9%で借入可能  
  - 変動金利だと **LIBOR + 1%**  
  - → 実際は **固定金利を希望**
- **投資銀行**：スワップを仲介し手数料を得る

---

### スワップの仕組み
1. **会社B と 投資銀行**
   - 投資銀行 → Bに **LIBORを支払う**  
   - B → 投資銀行に **7.5%支払う**  
   - B → 銀行に **LIBOR + 1%支払う**

   → 実質支払い = 7.5% + 1% = **8.5%**  
   （銀行の提示9%より有利）

2. **会社A と 投資銀行**
   - A → 投資銀行に **LIBOR支払う**  
   - 投資銀行 → Aに **7%支払う**  
   - A → 銀行に **固定6%支払う**

   → 実質負担 = LIBOR - 1%  
   （直接借入より1%有利）

3. **投資銀行**
   - Aから受け取ったLIBORをBへ渡す  
   - 差額 **(7.5% - 7% = 0.5%)** が収益  

---

### メリット
- **会社A**：LIBOR → LIBOR - 1% へ改善（1%得）  
- **会社B**：固定9% → 実質8.5%に改善（0.5%得）  
- **投資銀行**：差額で **0.5%の利益**

---

### まとめ
- 金利スワップは「固定金利を支払いたい企業」と「変動金利を支払いたい企業」の利害を調整し、投資銀行が仲介して全員が得をする仕組み。  
- 1980年代に広まり、現在も企業のリスク管理や資金調達に欠かせない金融手段。  


#5-4 クレジット・デフォルト・スワップ (CDS)

### CDSとは？
- **定義**：債務不履行（デフォルト）が起きた場合に、投資家（CDSの買い手）を補償する金融派生商品
- **仕組み**：  
  - CDS買い手 → 定期的にプレミアム（保険料のような支払い）を行う  
  - CDS売り手 → 万一、債券発行会社がデフォルトした場合、元本や利息を補償する  

---

### 背景
- 企業は資金調達のために **社債（Bonds）** を発行  
- 投資家（企業・年金基金など）は社債を購入 → 利息収入を得る  
- **リスク**：発行企業が倒産・債務不履行に陥る可能性  

---

### CDSの役割
- 投資家は「債務不履行リスク」に備えるためにCDSを購入  
- **保険のような仕組み**：
  - 通常時：投資家 → 投資銀行（CDS売り手）へ定期的に支払い  
  - デフォルト時：投資銀行（CDS売り手） → 投資家へ補償金を支払い  

---

### 関係者の役割
1. **債券発行会社（Borrower）**
   - 資金を得るために債券を発行
   - 利息と元本を返済する義務がある  

2. **投資家（Investor, CDS Buyer）**
   - 債券を購入し利息収入を得る  
   - デフォルトを懸念しCDSを購入  

3. **投資銀行（CDS Seller）**
   - CDSを販売してプレミアム収入を得る  
   - デフォルトが発生した場合は補償を行う  

---

### メリット
- **投資家**：デフォルトリスクを移転できる（保険効果）  
- **投資銀行**：デフォルト確率が低いと見込み、プレミアム収入を獲得  
- **市場全体**：信用リスクを「取引可能な商品」に変換できる  

---

### 問題点
- CDSが大量に取引されると、**システミックリスク**が拡大  
- 2008年金融危機では、CDSが過度に利用され危機を深刻化させた  
- 信用格付機関（S&P, Moody’s, JCR など）の評価が投資判断に大きな影響  

---

### まとめ
- CDSは「社債の信用リスクを保険化」する金融商品  
- 投資家はリスクを減らし、銀行はプレミアムで利益を得る  
- 2008年金融危機で注目を浴びた重要なデリバティブの1つ  


##5-5 オプション（Options）

### 定義
- **オプション**とは：  
  - 将来のある時点で、あらかじめ決められた価格で資産を売買する「権利」を与える金融派生商品  
  - ただし **義務はない**（取引するかどうかは投資家の自由）

---
### オプション取引の仕組み
1. 時刻 $t=0$ でオプションを購入（プレミアムを支払う）
2. 満期 $T$ で株価を確認
   - 株価が **上昇**していれば：安く買って高く売れる → 利益
   - 株価が **下落**していれば：権利を放棄 → 損失はオプション料のみ

---

### メリット
- **利益を狙える**：有利な場合のみ権利を行使できる
- **損失限定**：最大損失はオプション料（プレミアム）に限定される

---

### オプション価格の課題
- オプションの価値は **基礎資産（株価など）の変動**に依存
- 株価は常に変動するため、オプション価格も変動
- → 数理モデル（**ブラック＝ショールズモデル**）で価格を算出する必要がある

---

### 先物（Futures）との違い
| 項目     | 先物契約（Futures）       | オプション （Options）                |
| ------ | ------------------- | --------------------- |
| 契約内容   | 将来、決められた価格で**必ず売買** | 将来、決められた価格で売買する**権利** |
| 義務の有無  | **義務あり**            | **義務なし（権利のみ）**        |
| 不利な場合  | それでも**実行しなければならない** | **放棄できる**             |
| 有利な場合  | 実行する                | 実行する                  |
| 初期費用   | 原則なし（証拠金は必要）        | **プレミアム（料金）を支払う**     |
| 損失の大きさ | 理論上**無限大**の可能性      | **最大損失はプレミアムまで**      |
| 主な利用者  | 企業のヘッジ、投機家          | ヘッジ、投機、保険目的           |
| 例      | 航空会社が原油を将来**必ず買う**  | 株を将来**買ってもいい権利**      |




##5-6 コールオプション (Call Option)

### 定義
- **コールオプション**：将来、あらかじめ決められた価格で「株式を買う権利」
- 原資産（Underlying Asset）：通常は株式
- 義務はなく、**行使するかどうかは投資家が決められる**

---

### 主要なパラメータ
- **E (Exercise Price / Strike Price)**  
  行使価格。例：2,500円  
- **T (Expiry / Maturity)**  
  満期。例：1か月後  
- **S (Stock Price)**  
  株価。例：今日のトヨタ株 = 2,300円  
- **V (Value of Option)**  
  コールオプションの価格（プレミアム）

---

### 利用の目的
- 株価が **上昇する** と予想するときに購入する  
- 株価が上昇すれば、低い行使価格で株を買い、市場価格で売却 → **利益獲得**  
- 株価が行使価格を下回る場合は権利を行使せず、損失はプレミアムに限定  

---

### 満期時点での価値
コールオプションの価値は次の式で表される：

$$
V_T = \max(S_T - E, \; 0)
$$

- $S_T$：満期時の株価  
- $E$：行使価格  

---

### ペイオフ図（Payoff Diagram）
- 横軸：株価 $S_T$  
- 縦軸：オプション価値 $V_T$  
- 特徴：
  - $S_T < E$ のとき → $V_T = 0$（権利を行使しない）  
  - $S_T \geq E$ のとき → $V_T = S_T - E$（利益発生）  
  - 初期コスト（オプション料）を考慮すると、最初はマイナスからスタート  

---



```python
import numpy as np
import matplotlib.pyplot as plt

# パラメータ
S = np.linspace(0, 5000, 500)  # 満期時株価 S_T
E = 2500                       # 行使価格

# コールのペイオフ（プレミアムなし）
payoff_call = np.maximum(S - E, 0)

# 描画
plt.figure()
plt.plot(S, payoff_call, label="Call Option Payoff")
plt.axvline(E, linestyle="--", label="Strike Price E")
plt.xlabel("Stock Price at Maturity $S_T$")
plt.ylabel("Option Value $V_T$")
plt.title("Call Option Payoff Diagram")
plt.legend()
plt.grid(True)
plt.show()
```

---

### まとめ
- **コールオプション = 株を買う権利**
- 株価が行使価格を超えると利益が発生  
- 満期時は簡単な式で評価可能  
- 満期前の価格計算には **ブラック=ショールズモデル** が必要

##5-7 プットオプション (Put Option)

### 定義
- **プットオプション**：将来、あらかじめ決められた価格で「株式を売る権利」
- **コールオプション**：将来、あらかじめ決められた価格で「株式を買う権利」
- 注意：権利のみで義務はない

---

### 主要なパラメータ
- **E (Exercise Price / Strike Price)**  
  行使価格。例：2500円  
- **T (Expiry / Maturity)**  
  満期。例：1か月後  
- **S (Stock Price)**  
  株価。例：今日のトヨタ株 = 2500円
- **V (Value of Option)**  
  プットオプションの価格  

---

### 利用の目的
- 株価が **下落する** と予想するときに購入する  
- 株価が行使価格を下回れば、権利を行使して高値で売却 → **利益獲得**  
- 株価が行使価格を上回れば権利を放棄、損失はプレミアムに限定  

---

### 満期時点での価値
プットオプションの価値は次の式で表される：

$$
V_T = \max(E - S_T, \; 0)
$$

- $S_T$：満期時の株価  
- $E$：行使価格  

---

### ペイオフ図（Payoff Diagram）
- 横軸：株価 $S_T$
- 縦軸：オプション価値 $V_T$
- 特徴：
  - $S_T > E$ のとき → $V_T = 0$
  - $S_T \le E$ のとき → $V_T = E - S_T$


---


```python
import numpy as np
import matplotlib.pyplot as plt

# パラメータ
S = np.linspace(0, 5000, 500)  # 満期時株価 S_T
E = 2500                       # 行使価格

# プットのペイオフ
payoff_put = np.maximum(E - S, 0)

# 描画
plt.figure()
plt.plot(S, payoff_put, label="Put Option Payoff")
plt.axvline(E, linestyle="--", label="Strike Price E")
plt.xlabel("Stock Price at Maturity $S_T$")
plt.ylabel("Option Value $V_T$")
plt.title("Put Option Payoff Diagram")
plt.legend()
plt.grid(True)
plt.show()
```

---

### まとめ
- **プットオプション = 株を売る権利**
- 株価が行使価格を下回ると利益が発生  
- 満期時は簡単な式で評価可能  
- 満期前の価格計算には **ブラック=ショールズモデル** が必要

##5-8 アメリカンオプションとヨーロピアンオプション

### アメリカンオプション (American Option)
- 満期日 **以前の任意の時点で行使可能**
- 柔軟性が高い
- 例：株価が有利に動いた時点で即行使できる

---

### ヨーロピアンオプション (European Option)
- **満期日 (T)** にのみ行使可能
- より単純でモデル化しやすい
- ブラック=ショールズモデルは **ヨーロピアンオプションを前提**

---

### まとめ
- アメリカン: 「いつでも行使できる」  
- ヨーロピアン: 「満期日にしか行使できない」  
- 数理モデル上はヨーロピアンの方が扱いやすい  
- 実務上はアメリカンの方が柔軟だが、評価が難しい

#6 ファイナンスにおけるランダム性

#6-1分析手法の種類

3つの主要な分析手法

---

##1. ファンダメンタル分析（Fundamental Analysis）

企業そのものを詳しく調査し、本質的価値（Intrinsic Value）に対して株価が割安かどうかを判断する手法。

#### 主な分析対象
- 経営陣
- 製品・サービス
- 資産（現金・設備など）
- 負債（借金・ローン）
- 損益計算書（収益・費用）
- 貸借対照表
- キャッシュフロー計算書

#### 重要指標：PER（株価収益率）
-  純利益 ÷ 発行済株式数 = EPS  
-  株価 ÷ EPS = PER  → **低いほど割安**と判断されやすい



#### 特徴
- 本質的価値を重視  
- 需給がすぐに反映されない場合、正しく割安でも株価が上がらないこともある  
- 長期投資向けの手法

---

##2. テクニカル分析（Technical Analysis）

企業の情報は株価にすべて反映されていると仮定し、**過去の価格データから未来を予測する手法**。

#### 主な手法
- 移動平均線（SMA）
- RSI（相対力指数）
- AR / ARIMA モデル
- ロジスティック回帰、SVM、DNN などの機械学習モデル

#### 基本概念
> 過去に見られたパターンは将来も繰り返される  
→ 過去データから未来の動きを予測する

#### 特徴
- チャートの形やパターンを重視  
- 企業の財務情報は直接見ない  
- 短期〜中期取引に有効

---

##3. クオンツ分析（Quantitative Analysis）

高度な数学と確率論を用いて、金融商品のランダムな動きをモデル化する手法。

#### 必要となる数学
- 統計学
- 確率論
- 確率過程
- 確率微分方程式（SDE）
- 確率解析

#### 特徴
- 資産のランダム性を前提とする  
- ポートフォリオ最適化などでリスクを最小化  
- モデル例：ブラック＝ショールズ方程式（オプション価格計算）

---

### まとめ

| 手法 | 何を見る？ | 主な特徴 |
|------|------------|-----------|
| ファンダメンタル分析 | 企業の価値・財務 | 割安/割高を判断、長期向け |
| テクニカル分析 | 過去の価格データ | パターン分析で未来を予測、短期向け |
| クオンツ分析 | 数学モデル・確率過程 | ランダム性を数式化、高度で精密 |



#6-2 資産のランダム性（確率的挙動）

金融資産が「ランダムに振る舞う」という前提の背景と、その理由について解説する。

---

##1. なぜランダム性を仮定するのか？

クオンツ分析では、**株価・金利・オプション価格などはランダムに動く**と仮定する。

その理由は、実際の市場データ（特に**日次リターン**）を調べると、以下のような特徴があるため：

- 日次リターン = (今日の株価 − 昨日の株価) ÷ 昨日の株価
- 日次リターンのヒストグラムを描画すると  → **平均0の正規分布に非常に近い形状**になる
- 月次リターンでも同様の傾向がある

正規分布は **「平均」「分散（または標準偏差）」** の 2 つのパラメータで特徴付けられる。

標準偏差 = 分散の平方根。→ **リターンは正規分布からの乱数とみなせる**  という結論に至る。

---

##2. リターンをランダム変数として表す

リターン $ R $ は次のように表せる：

$$
R = \mu + \sigma \varepsilon
$$

- $ \mu $：平均
- $ \sigma $：標準偏差
- $ \varepsilon $：正規分布からの乱数（$\varepsilon \sim \mathcal{N}(0,1)$）

これは日次リターンが正規分布に従うという観察に基づいている。

---

##3. 正規分布がよく現れる理由（中心極限定理）

複数の独立な確率変数を足し合わせると、その和は正規分布に近づく。

これが **中心極限定理（Central Limit Theorem）**。

そのため金融市場だけでなく、物理・統計などあらゆる分野で正規分布が頻出する。

---

##4. 株価の変動もランダム変数で表す

リターンだけでなく **株価そのものの未来値もランダム変数** として扱える。

例：

- 今日の株価：$ S_t $
- 明日の株価：$ S_{t+1} $  
  → **正規分布からの乱数**として扱える

これが **ウィーナー過程（Wiener Process）** を使ったモデル化の考え方につながる。

---

##5. ウィーナー過程の特徴

ウィーナー過程は：

- **連続的なパスを持つ**
- **増分が独立で、かつ正規分布に従う**

株価の変化をこのプロセスでモデル化するのは、「日次リターンが正規分布に従う」という経験的事実に基づく。

---

##6. 株価の大まかな振れ幅

将来の株価は不確実だが、次の範囲に収まる可能性が高い：

$$
\mu \pm 1 \sigma
$$

- 平均（トレンド）  
- ± 標準偏差（ばらつき）

多くの場合、株価はこのレンジ内で推移する。

---

##7. 今日の株価から明日の株価を予測する考え方

今日の株価を $ S_t $ とすると、明日の株価 $ S_{t+1} $ は：

$$
ln \frac{S_{t+1}}{S_t} \sim \mathcal{N}(\mu, \sigma^2 )
$$
よって、
$$S_{t+1} = S_t exp(\mu +\sigma\varepsilon)$$
つまり 株価は正規分布ではなく、対数正規分布に従う。これがクオンツ分析の基本的な前提となる。



#6-3 金融資産の数学的背景：ウィーナー過程とランダムウォーク

金融資産の確率的挙動を説明するための **ウィーナー過程** と **ランダムウォーク（特に幾何ブラウン運動）** について解説する。

---
##1. 株価そのものは正規分布に従わない理由

-  理由1：株価は負の値を取れない  
正規分布は負の値が出る → モデルとして破綻。

- 理由2：株価には長期的な上昇トレンドがある  
Apple、Tesla、Google などの株価は **長期では上昇し続ける傾向** にある。

  → 完全なランダムウォーク（通常の正規分布）では説明できない。

---

## 2. 株価は対数正規分布に従う

クオンツ分析では、株価 $S_t$ そのものではなく、  
**株価の対数変化** を確率変数としてモデル化する。

今日の株価を $S_t$ とすると、  
明日の株価 $S_{t+1}$ に対する **対数リターン** は次のように表される：

$$
\ln \frac{S_{t+1}}{S_t} \sim \mathcal{N}(\mu, \sigma^2)
$$

したがって、株価の更新式は：

$$
S_{t+1} = S_t \exp(\mu + \sigma \varepsilon),
\quad \varepsilon \sim \mathcal{N}(0,1)
$$

このとき、株価 $S_{t+1}$ は  
**正規分布ではなく対数正規分布（lognormal distribution）** に従う。

この仮定により：

- 株価は常に正の値を取る  
- 長期的に指数関数的な成長が可能  
- 株価が大きくなるほど変動も大きくなる  

といった、実際の市場の特徴を自然に説明できる。

このモデル化が、クオンツ分析における  **最も基本的な前提のひとつ** である。


---

##3. 株価変動は「トレンド + ノイズ」
$$
S_{t+1} = S_t \exp(\mu + \sigma \varepsilon)
$$

というモデルでは、株価の変動は **決定論的な成分** と **確率的な成分** に分けて解釈できる。

株価の動きは以下の 2 つの要素から構成される：

1. **決定論的部分（ドリフト）**  
   → 長期的な平均成長率を表すパラメータ $\mu$

2. **確率的部分（ランダム変動）**  
   → 価格変動の大きさを表すパラメータ $\sigma$ （ボラティリティ）


このように、株価変動は「トレンド + ノイズ」として捉えられる。


---

##4. ウィーナー過程の性質
ここで問題となるのは、

> この「ランダムなノイズ」を   どのような確率過程で表すか

である。

クオンツ分析では、この確率的な揺らぎを **ウィーナー過程（Wiener process）**  によってモデル化する。

ウィーナー過程 $W_t$ の増分は、次の性質を持つ：

$$
W_{t+\Delta t} - W_t \sim \mathcal{N}(0, \Delta t)
$$

- 増分は互いに独立  
- 増分は正規分布に従う  
- サンプルパスは連続  

このように、ウィーナー過程は「平均0の純粋なノイズ」を表す確率過程であり、株価変動のランダム成分を記述するための基礎となる。

---

##5. 幾何ブラウン運動（Geometric Brownian Motion: GBM）

金融工学で最も重要なSDE（確率微分方程式）はこれ：

$$ dS_t = \mu S_t \, dt + \sigma S_t \, dW_t $$

意味：

- $dS_t$：株価の微小変化  
- $\mu S_t dt$：決定論的トレンド  
- $\sigma S_t dW_t$：確率的変動（ボラティリティ）  
- $dW_t$：正規分布 $\mathcal{N}(0, dt)$ に従う増分
---
##6. 株価が大きいほど変動が大きくなる理由

幾何ブラウン運動

$$
dS_t = \mu S_t \, dt + \sigma S_t \, dW_t
$$

において、確率的変動を表す項は  
$\sigma S_t dW_t$ である。

この式から分かるように、株価の変動幅は

$$
\text{変動幅} \propto S_t
$$

となる。

すなわち、株価が大きくなるほど  
絶対的な変動も自然に大きくなる。

これは、実際の市場において  
高価格帯の株ほど値動きが大きくなるという  
経験的事実と一致している。

---

##7. 現代金融モデルへの応用

この幾何ブラウン運動

$$
dS_t = \mu S_t \, dt + \sigma S_t \, dW_t
$$

は、現代金融工学の中心的なモデルであり、

- ブラック＝ショールズモデル  
- 金利モデル（Vasicek、CIR など）  
- 各種デリバティブの価格付け理論  

といった多くの理論の基盤となっている。



#6-4 ウィーナー過程とランダムウォークのシミュレーション
##1. 株価そのものではなく「過程」をモデル化する理由

株価の動きを連続時間で変化する確率的な対象として扱うため、クオンツ分析では **確率過程** を用いる。

その代表例が：

- ランダムウォーク（離散時間モデル）  
- ウィーナー過程（連続時間モデル）  

である。

ウィーナー過程（ブラウン運動）は、ランダムウォークを時間刻みを無限に小さくした**連続時間極限** として定義される。

ウィーナー過程 $W(t)$ は次の性質を持つ：

- 増分は独立  
- 増分は正規分布に従う  

すなわち、任意の $\Delta t > 0$ に対して：

$$
W(t+\Delta t) - W(t) \sim \mathcal{N}(0, \Delta t)
$$

---


## 2. 数値シミュレーションの基本アイデア

連続時間の確率過程は、そのままではコンピュータ上で扱えない。  
そこで、小さな時間刻み $dt$ を用いて近似する。

基本的な手順は次の通り：

1. 標準正規乱数 $Z \sim \mathcal{N}(0,1)$ を生成  
2. ウィーナー過程の増分を  
   $$
   dW = \sqrt{dt} \times Z
   $$
   として計算  
3. 初期値 $W(0)=0$ を与え、増分を累積することで  
   $W(t)$ のパスを構築する  

この方法により、ウィーナー過程のサンプルパスを数値的にシミュレーションできる。

---

## 3. Python実装：ウィーナー過程クラス

以下に示す **WienerProcess** クラスは、ウィーナー過程の生成と可視化を目的とする。

### 特徴
- 時間刻み $dt$ とステップ数 $N$ を指定  
- `generate_path`：ウィーナー過程のパスを生成  
- `plot_path`：生成したパスを可視化  
---

## 4.Python実装例

```python
import numpy as np
import matplotlib.pyplot as plt

class WienerProcess:
    def __init__(self, dt=0.1, N=1000, seed=None):
        """
        dt : 時間刻み
        N  : ステップ数
        seed : 乱数シード（再現性のため、任意）
        """
        self.dt = float(dt)
        self.N = int(N)
        self.rng = np.random.default_rng(seed)

        self.t = None
        self.W = None

    def generate_path(self):
        """
        ウィーナー過程 W(t) を生成して内部に保持する
        W(0)=0,  dW ~ N(0, dt)
        """
        # 時間軸
        self.t = np.linspace(0.0, self.N * self.dt, self.N + 1)

        # 標準正規乱数 Z ~ N(0, 1)
        Z = self.rng.standard_normal(self.N)

        # 増分 dW = sqrt(dt) * Z  → dW ~ N(0, dt)
        dW = np.sqrt(self.dt) * Z

        # 累積和で W を構築（W[0]=0）
        self.W = np.empty(self.N + 1, dtype=float)
        self.W[0] = 0.0
        self.W[1:] = np.cumsum(dW)

        return self.t, self.W

    def plot_path(self):
        """
        生成されたパスをプロットする
        """
        if self.W is None or self.t is None:
            self.generate_path()

        plt.figure(figsize=(10, 5))
        plt.plot(self.t, self.W, label="Wiener Process")
        plt.xlabel("Time")
        plt.ylabel("W(t)")
        plt.title("Wiener Process Simulation")
        plt.grid(True)
        plt.legend()
        plt.show()


if __name__ == "__main__":
    process = WienerProcess(dt=0.1, N=1000, seed=42)
    process.plot_path()
```

---
## 5. 株価モデルとしての評価

ウィーナー過程には次のような特徴（欠点）がある：

- 値が正にも負にもなり得る  
- 長期的なトレンドは 0  
- ランダムでギザギザした動きをする  

一方、実際の株価には次の性質がある：

- 負の値を取らない（最小値は 0）  
- 長期的に上昇トレンドを持つ  
- 株価が高いほど変動（ボラティリティ）も大きくなる  

したがって、**ウィーナー過程そのものは株価モデルとして不適切である。**

---

## 6. 幾何ブラウン運動（GBM）への接続

株価をモデル化するためには、  
次の確率微分方程式（SDE）が広く用いられる：

$$
dS(t) = \mu S(t)\,dt + \sigma S(t)\,dW(t)
$$

### 特徴

- 決定論的トレンド $\mu$（期待リターン）  
- 確率的変動 $\sigma$（ボラティリティ）  
- 解が指数関数形となり、株価は常に正  
- 現代金融工学における標準的な株価モデル  

---

## 7. まとめ

- 対数リターンは正規分布で近似できる  
- ウィーナー過程は最も基本的な確率過程  
- numpy の累積和で容易にシミュレーション可能  
- しかし株価を表すには不十分（負値・トレンドがない）  
- そのため、**幾何ブラウン運動（GBM）** が株価モデルとして用いられる  


#6-5 伊藤計算と確率変数の微分
「確率微分方程式（SDE）を解くための伊藤の補題」について整理する。

---
## 1. 普通の関数（確定変数）の微分

まず、確定変数 $x$ をもつ通常の関数を考えます。

例：  
$$f(x) = x^2$$

この微分は：

$$\frac{df}{dx} = 2x$$

小さな変化 $dx$ に対する変化は、

$$df = 2x\,dx$$

これは普通の微分の公式であり、関数が確定的で、$x$ がランダムでない場合に成り立つ。

---

### 2. 確率変数（ウィーナー過程）を含む場合はどうなるか？

今度は $x$ が **確率変数（例：ウィーナー過程 $W_t$）** の場合を考える。

ウィーナー過程の性質：

- $dW$ は平均 0、分散 $dt$ の正規分布  
- つまり  
  $dW \sim N(0, dt)$

ここで問題：

> 普通の微分法則 $df = f'(x)\,dx$ はそのまま使えるか？

→ **使えない。**

理由：  
$W_t$ は確率変数であり、$dW$ は無限小ではあるがランダムで、  
しかも二乗すると $dW^2 = dt$ という非常に特殊な性質を持つため。

---

### 3. 伊藤の補題

確率変数を含む関数 $f(x)$ の変化は次で与えられる：

$$df = f'(x)\,dx + \frac{1}{2} f''(x)\,(dx)^2$$

ここで、ウィーナー過程の性質

- $dW$ の期待値は 0
- $(dW)^2 = dt$

を使う。

ただの微分との違い：

- **2次の微小項 $(dx)^2$ が消えない**
- むしろ $(dW)^2 = dt$ となって残る

これが確率微分の最大のポイント。

---

### 4. Taylor 展開からの導出（直感）

関数 $f(x)$ を $x + dx$ で評価すると：

$$f(x + dx) = f(x)  
\quad + f'(x)\,dx  
\quad + \frac{1}{2} f''(x)\,(dx)^2  
\quad + \cdots$$

通常の微分では $(dx)^2$ は無視してよい小ささだが、  
確率変数では **無視すると間違いになる**。

なぜか？

- $dx = dW$ の場合  
- $dW \sim \sqrt{dt}$ の大きさ  
- $(dW)^2 = dt$ は無視できない

---

### 5. 実際に計算してみる：$f(x)=x^2$

#### 関数
$$f(x) = x^2$$

#### 1次微分
$$f'(x) = 2x$$

#### 2次微分
$$f''(x) = 2$$

#### 伊藤の補題を適用

$$df = 2x\,dW + \frac{1}{2}\cdot 2 \cdot (dW)^2$$

$$dW^2 = dt$$ を使って簡略化：

$$df = 2x\,dW + dt$$

---

### 6. 普通の微分との比較

確定変数の場合：

$$df = 2x\,dx$$

確率変数の場合（伊藤の補題）：

$$df = 2x\,dW + dt$$

**→ “+ dt” が追加されるのが決定的な違い。**

この追加項が後の

- 幾何ブラウン運動（GBM）
- Black–Scholes 方程式の導出
- 金融工学の確率微分方程式

すべてに現れる。

---

### 7. まとめ

- 普通の微分では 1次の微小項だけが残る  
- ウィーナー過程では $(dW)^2 = dt$ が残るため 2次項が重要  
- これを体系化したのが **伊藤の補題**  
- 金融工学では必須の道具  
- 次のステップは **幾何ブラウン運動（GBM）に伊藤の補題を適用すること**





##6-6 高次元伊藤の補題とブラック–ショールズ方程式の導入

この資料では、講義で説明されていた  
**「なぜ確率微分方程式（SDE）と伊藤の補題が必要なのか」**  
を整理しつつ、ブラック–ショールズ式につながる *多変数の伊藤の補題* をまとめる。

---

### 1. なぜ確率微分方程式（SDE）が必要なのか？

オプション価格 $V(S,t)$ を計算したいとき：

- 満期（$t = T$）ではオプションの価値は明確  
  例：コールなら $V = \max(S_T - K, 0)$

- しかし知りたいのは **満期前 $t < T$ のオプション価格**  
  → この時点では株価 $S_t$ が確率的に変動するため、  
     オプション価格 $V(S,t)$ も確率的に変動する

よって、株価と同じく **確率過程としてオプション価格を扱う必要がある**。

そのために **確率微分方程式(SDE)** と **伊藤の補題** が必要になる。

---

### 2. 株価のモデル：SDE（一次元）

株価の変化は次の形でモデル化されたとする：

$$ dS = a\,dt + b\,dW $$

- $a = \mu S$（ドリフト：期待成長率）
- $b = \sigma S$（ボラティリティ）

ただし今回は記号を簡略化して $a, b$ をそのまま使う。

---

### 3. オプション価格は 2 変数の確率関数 になる

オプション価格 $V$ は次を変数に持つ：

- 株価：$S$（確率変数）
- 時間：$t$（確定変数）

つまり：

$$ V = V(S, t) $$

変化量は：

$$ dV = V_S\,dS + V_t\,dt + \frac{1}{2}V_{SS}(dS)^2 $$

多変数の伊藤の補題である。

---

### 4. Taylor 展開からの導出のイメージ

$V(S,t)$ を $(S + dS, t + dt)$ に広げると：

$$ V(S+dS, t+dt) =  
V + V_S dS + V_t dt + \frac{1}{2}V_{SS}(dS)^2 + \cdots $$

通常の微分なら $(dS)^2$ は無視して良いが、  
確率変数では **無視すると間違い**。

理由：  
$dS = a\,dt + b\,dW$ なので：

$$(dS)^2 = (a\,dt + b\,dW)^2  
\approx b^2(dW)^2 = b^2\,dt$$

ここでウィーナー過程の性質：

$$(dW)^2 = dt$$

が重要。

---

### 5. 必要な項だけ残すと…

$dS = a\,dt + b\,dW$ を使って：

$$ (dS)^2 \approx b^2 dt $$

これを伊藤の式に代入：

$$ dV  
= V_t dt  
+ V_S (a\,dt + b\,dW)  
+ \frac{1}{2} V_{SS} b^2 dt $$

まとめると：

$$ dV  
= \left( V_t + a V_S + \frac{1}{2}b^2 V_{SS} \right) dt  
+ b V_S\, dW $$

これは **オプション価格の確率微分方程式** であり、  
後にブラック–ショールズ PDE に変形される。

---

### 6. なぜこれが重要なのか？

株価 $S$ が確率過程である以上：

- $S$ 依存のオプション価格 $V(S,t)$ も確率過程になる  
- よって $V$ の変化量 $dV$ を正しく扱うには  
  **高次元の伊藤補題** が不可欠

ブラック–ショールズの導出では：

- この式をリスクヘッジ資産に組み合わせることで  
- $dW$ の項を消し、リスクのないポートフォリオを作り  
- PDE（偏微分方程式）へ変形する

という流れになる。

---

### 7. まとめ

- 株価が確率過程 → オプション価格も確率過程  
- よって 2 変数の伊藤補題が必要  
- $dV$ には 1次項（$V_S$, $V_t$）だけでなく  
  **2次項の $\frac{1}{2}V_{SS}(dS)^2$ が必須**
- $(dS)^2 = b^2 dt$ という性質に注意  
- これが後のブラック–ショールズ方程式の核心部分



##6-7 幾何ブラウン運動（GBM）の確率微分方程式を伊藤の補題で解く

この資料では、講義で説明されていた  
**「幾何ブラウン運動（GBM）の SDE を伊藤の補題で解く手順」**  
をわかりやすく整理したものです。

---

### 1. 幾何ブラウン運動（GBM）の定義

株価 $S(t)$ の変化は次の確率微分方程式（SDE）で与えられる：

$$ dS = \mu S\,dt + \sigma S\,dW $$

- $\mu$：ドリフト（平均成長率）
- $\sigma$：ボラティリティ
- $W$：ウィーナー過程（ブラウン運動）

この式を **解いて明示的な形の $S(t)$** を得るのが今回の目的。

---

### 2. まず式を「対数微分の形」にする

両辺を $S$ で割る：

$$ \frac{dS}{S} = \mu\,dt + \sigma\,dW $$

これは「対数微分」と同じ形なので：

$ d(\ln S)$ を伊藤の補題で計算すれば良い。

よって、次の関数を定義する：

$$f(S) = \ln S$$

---

### 3. 伊藤の補題（Itô's Lemma）を適用

伊藤の補題（1変数版）は：

$$ df = f'(S)\,dS + \frac{1}{2} f''(S)\,(dS)^2 $$

ここで：

#### 1次微分
$$f'(S) = \frac{1}{S}$$

#### 2次微分
$$f''(S) = -\frac{1}{S^2}$$

---

### 4. $(dS)^2$ を計算する

$$ dS = \mu S\,dt + \sigma S\,dW $$

の両辺を二乗する：

$$(dS)^2  
= (\mu S\,dt + \sigma S\,dW)^2  
\approx \sigma^2 S^2 (dW)^2  
= \sigma^2 S^2 dt $$

理由：$dt^2$ や $dt\,dW$ は無視でき、$(dW)^2 = dt$ のみ残る

---

### 5. 伊藤の補題に代入

$$df = f'(S)\,dS + \frac{1}{2}f''(S)(dS)^2$$

代入：

$$df  
= \frac{1}{S}(\mu S\,dt + \sigma S\,dW)  
\quad + \frac{1}{2}\left(-\frac{1}{S^2}\right)(\sigma^2 S^2 dt)$$

簡略化：

$$df  
= \mu\,dt + \sigma\,dW  
\quad - \frac{1}{2}\sigma^2 dt$$

よって：

$$ d(\ln S)  
= \left(\mu - \frac{1}{2}\sigma^2\right) dt + \sigma\, dW $$

---

### 6. 両辺を積分して明示解を得る

0〜t まで積分：

$$\ln S(t) - \ln S(0)  
= \left(\mu - \frac{1}{2}\sigma^2\right)t + \sigma W(t)$$

両辺に指数をつけて $S(t)$ を求める：

$$ S(t)  
= S(0)\,\exp\left[\left(\mu - \frac{1}{2}\sigma^2\right)t + \sigma W(t)\right] $$

---

### 7. 幾何ブラウン運動の解の特徴

- 常に正の値をとる（指数関数形のため）
- 平均成長率 $\mu$ に従って長期的に増えていく
- ボラティリティ $\sigma$ が大きいほど変動も大きい
- 株価の現実的特徴を非常によく表す

→ この理由で **GBM は現代金融工学の標準的な株価モデル** になっている。

---

### 8. まとめ

- SDE $dS=\mu S dt + \sigma S dW$ を伊藤の補題で解いた  
- 対数変換 $f(S)=\ln S$ を使うのがポイント  
- $(dW)^2 = dt$ を利用して 2次項が残る  
- 最終解は：

$$ S(t) = S(0)\exp\big[(\mu - \tfrac{1}{2}\sigma^2)t + \sigma W(t)\big] $$

- これが次の講義でシミュレーションする **GBM の明示解** となる



##6-8 幾何ブラウン運動（GBM）のPython実装

「幾何ブラウン運動（ GBM）のシミュレーション」
を、Python 実装してまとめる。

---

###1. GBM シミュレーションのアルゴリズム

1. 時間刻み  
   $$dt = \frac{T}{N}$$ を計算する  
2. 時間軸  
   `np.linspace(0, T, N+1)`  
3. 標準正規乱数を生成  
   $$Z \sim \mathcal{N}(0,1)$$
4. $W(t)$ を累積和で生成  
   $$W(t) = \sqrt{dt} \sum Z$$  
5. 明示解  
   $$S(t) = S_0 \exp[(\mu - \tfrac12\sigma^2)t + \sigma W(t)]$$

---

###2.Python コード：幾何ブラウン運動（GBM）の実装

  

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

# =========================
#  ウィーナー過程クラス
# =========================
class WienerProcess:
    def __init__(self, dt=0.1, N=1000):
        """
        初期化
        dt : 時間刻み
        N  : ステップ数
        """
        self.dt = dt
        self.N = N
        self.t = None
        self.W = None

    def generate_path(self):
        """ウィーナー過程 W(t) を生成して内部に保持する"""
        # 時間軸
        self.t = np.linspace(0, self.N * self.dt, self.N + 1)

        # 標準正規乱数 Z ~ N(0, 1)
        Z = npr.normal(0, 1, self.N)

        # 増分 dW = sqrt(dt) * Z
        dW = np.sqrt(self.dt) * Z

        # 初期値 0 の配列
        self.W = np.zeros(self.N + 1)

        # 累積和で W を構築
        self.W[1:] = np.cumsum(dW)

        return self.t, self.W

    def plot_path(self):
        """生成されたパスをプロットする"""
        if self.W is None:
            self.generate_path()

        plt.figure(figsize=(10, 5))
        plt.plot(self.t, self.W, label="Wiener Process")
        plt.xlabel("Time")
        plt.ylabel("W(t)")
        plt.title("Wiener Process Simulation")
        plt.grid(True)
        plt.legend()
        plt.show()


# =========================
#  幾何ブラウン運動 (GBM)
# =========================
def simulate_gbm(S0, T, N, mu, sigma):
    """
    幾何ブラウン運動 (GBM) のシミュレーション
    S0   : 初期株価
    T    : シミュレーション期間（年）
    N    : ステップ数
    mu   : ドリフト（平均成長率）
    sigma: ボラティリティ
    """
    # 1. 時間刻み
    dt = T / N

    # 2. 時間軸
    t = np.linspace(0, T, N + 1)

    # 3. 標準正規乱数
    Z = np.random.normal(0, 1, N)

    # 4. Brownian motion
    W = np.zeros(N + 1)
    W[1:] = np.cumsum(np.sqrt(dt) * Z)

    # 5. GBM の明示解
    S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

    return t, S


def plot_gbm(t, S):
    plt.figure(figsize=(10, 5))
    plt.plot(t, S, label="Geometric Brownian Motion")
    plt.xlabel("Time")
    plt.ylabel("Stock Price")
    plt.title("GBM Simulation")
    plt.grid(True)
    plt.legend()
    plt.show()


# =========================
#  メイン処理（ノートブックではそのまま実行）
# =========================
if __name__ == "__main__":
    # ウィーナー過程
    process = WienerProcess(dt=0.1, N=1000)
    process.plot_path()

    # GBM
    S0 = 10       # 初期株価
    T = 2         # 2年間
    N = 1000      # 1000ステップ
    mu = 0.1      # 平均成長率（10%）
    sigma = 0.3   # ボラティリティ

    t, S = simulate_gbm(S0, T, N, mu, sigma)
    plot_gbm(t, S)
```

---

###3. 出力される株価軌道の特徴

シミュレーション結果を確認すると：

- 上昇トレンド（μによる）
- ランダム変動（σによる）
- 価格は常に 0 より大きい → 株価の性質に合致  
- σ を大きくすると変動が大きくなる
- S0 を変えると開始点が変わる

たとえば：

- `sigma = 0.05` → 滑らかで安定  
- `sigma = 0.5` → 激しく上下に揺れる  

---

###4. このモデルが株価に適している理由

幾何ブラウン運動 GBM は株価の性質を自然に再現する：

- 負の値を取らない（下限が 0）
- 長期的に増加する傾向（経済成長）
- 価格が大きいと変動幅も大きくなる（対数正規性）
- 金融工学で最も広く使われる標準モデル

（Black–Scholes 価格公式も GBM を前提に導かれる）

---

###5. まとめ

- GBM の SDE：$dS = \mu S dt + \sigma S dW$  
- 伊藤の補題で解くと明示式が得られる  
- Brownian motion を累積和で生成し、指数に代入するだけでシミュレーション可能  
- Python では NumPy を使って効率的に実装できる  
- 株価の現実的特徴（対数正規性・非負性）を自然に満たす

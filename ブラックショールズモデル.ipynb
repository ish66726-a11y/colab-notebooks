#11-1分析手法の種類

この講義では、3つの主要な分析手法について解説しています。

---

## 1. ファンダメンタル分析（Fundamental Analysis）

企業そのものを詳しく調査し、本質的価値（Intrinsic Value）に対して株価が割安かどうかを判断する手法。

### 主な分析対象
- 経営陣
- 製品・サービス
- 資産（現金・設備など）
- 負債（借金・ローン）
- 損益計算書（収益・費用）
- 貸借対照表
- キャッシュフロー計算書

### 重要指標：P/E（株価収益率）
1. 純利益 ÷ 発行済株式数 = EPS  
2. 株価 ÷ EPS = P/E比率  
→ **低いほど割安**

### 特徴
- 本質的価値を重視  
- 需給がすぐに反映されない場合、正しく割安でも株価が上がらないこともある  
- 長期投資向けの手法

---

## 2. テクニカル分析（Technical Analysis）

企業の情報は株価にすべて反映されていると仮定し、**過去の価格データから未来を予測する手法**。

### 主な手法
- 移動平均線（SMA）
- RSI（相対力指数）
- AR / ARIMA モデル
- ロジスティック回帰、SVM、DNN などの機械学習モデル

### 基本概念
> 過去に見られたパターンは将来も繰り返される  
→ 過去データから未来の動きを予測する

### 特徴
- チャートの形やパターンを重視  
- 企業の財務情報は直接見ない  
- 短期〜中期取引に有効

---

## 3. クオンツ分析（Quantitative Analysis）

高度な数学と確率論を用いて、金融商品のランダムな動きをモデル化する手法。

### 必要となる数学
- 統計学
- 確率論
- 確率過程
- 確率微分方程式（SDE）
- ストキャスティック・カルキュラス（確率解析）

### 特徴
- 資産のランダム性を前提とする  
- ポートフォリオ最適化などでリスクを最小化  
- モデル例：ブラック＝ショールズ方程式（オプション価格計算）

---

## まとめ

| 手法 | 何を見る？ | 主な特徴 |
|------|------------|-----------|
| ファンダメンタル分析 | 企業の価値・財務 | 割安/割高を判断、長期向け |
| テクニカル分析 | 過去の価格データ | パターン分析で未来を予測、短期向け |
| クオンツ分析 | 数学モデル・確率過程 | ランダム性を数式化、高度で精密 |

#11-2 資産のランダム性（確率的挙動）

金融資産が「ランダムに振る舞う」という前提の背景と、その理由について解説する。

---

## 1. なぜランダム性を仮定するのか？

クオンツ分析では、**株価・金利・オプション価格などはランダムに動く**と仮定する。

その理由は、実際の市場データ（特に**日次リターン**）を調べると、以下のような特徴があるため：

- 日次リターン = (今日の株価 − 昨日の株価) ÷ 昨日の株価
- 日次リターンのヒストグラムを描画すると  
  → **平均0の正規分布に非常に近い形状**になる
- 月次リターンでも同様の傾向がある

正規分布は  
**「平均」「分散（または標準偏差）」** の 2 つのパラメータで特徴付けられる。

標準偏差 = 分散の平方根。

→ **リターンは正規分布からの乱数とみなせる**  
という結論に至る。

---

## 2. リターンをランダム変数として表す

リターン $( R $) は次のように表せる：

$[
R = \mu + \sigma \varepsilon
$]

- $( \mu $)：平均（決定論的部分）
- $( \sigma $)：標準偏差
- $( \varepsilon $)：正規分布からの乱数（ランダム部分）

これは日次リターンが正規分布に従うという観察に基づいている。

---

## 3. 正規分布がよく現れる理由（中心極限定理）

複数の独立な確率変数を足し合わせると、その和は正規分布に近づく。

これが **中心極限定理（Central Limit Theorem）**。

そのため金融市場だけでなく、物理・統計などあらゆる分野で正規分布が頻出する。

---

## 4. 株価の変動もランダム変数で表す

リターンだけでなく **株価そのものの未来値もランダム変数** として扱える。

例：

- 今日の株価：$( S_t $)
- 明日の株価：$( S_{t+1} $)  
  → **正規分布からの乱数**として扱える

これが **ウィーナー過程（Wiener Process）** を使ったモデル化の考え方につながる。

---

## 5. ウィーナー過程（Wiener Process）の特徴

ウィーナー過程（ブラウン運動）は：

- **連続的なパスを持つ**
- **増分が独立で、かつ正規分布に従う**

株価の変化をこのプロセスでモデル化するのは、  
「日次リターンが正規分布に従う」という経験的事実に基づく。

---

## 6. 株価の大まかな振れ幅

将来の株価は不確実だが、次の範囲に収まる可能性が高い：

$[
\mu \pm 1 \sigma
$]

- 平均（トレンド）  
- ± 標準偏差（ばらつき）

多くの場合、株価はこのレンジ内で推移する。

---

## 7. 今日の株価から明日の株価を予測する考え方

今日の株価を $( S_t $) とすると、明日の株価 $( S_{t+1} $) は：

$[
S_{t+1} \sim \text{Normal}(\text{mean}, \text{variance})
$]

つまり **株価の将来値そのものがランダムウォークである**。

これがクオンツ分析の基本的な前提となる。



#11-3 金融資産の数学的背景：ウィーナー過程とランダムウォーク

この講義では、金融資産の確率的挙動を説明するための **ウィーナー過程（Wiener Process）** と  
**ランダムウォーク（特に幾何ブラウン運動）** について解説する。

---

## 1. リターンは正規分布に従う

株価 $S_t$ を使うと **日次リターン** は次の式で計算できる：

$ \text{Daily Return} = \frac{S_t - S_{t-1}}{S_{t-1}} $

日次リターンのヒストグラムは：

- 平均はほぼ 0  
- 一定の分散（標準偏差）  
- 正規分布に非常に近い形状  

→ **リターンは正規分布に従う** とみなせる。

---

## 2. 株価そのものは正規分布に従わない理由

### 理由1：株価は負の値を取れない  
正規分布は負の値が出る → モデルとして破綻。

### 理由2：株価には長期的な上昇トレンドがある  
Apple、Tesla、Google などの株価は **長期では上昇し続ける傾向** にある。

→ 完全なランダムウォーク（通常の正規分布）では説明できない。

---

## 3. 株価は対数正規分布に従う

株価 $S_t$ ではなく **その対数 $\ln S_t$ が正規分布に従う** と仮定する。

$ \ln S_t \sim \mathcal{N}(\mu, \sigma^2) $

つまり株価そのものは **対数正規分布（lognormal distribution）**。

この仮定により：

- 株価は負にならない  
- 長期的に指数関数的な増加が可能  
- 株価が大きくなると変動も大きくなる（実際の市場と一致）

---

## 4. 株価変動は「トレンド + ノイズ」

株価の動きは以下に分解できる：

1. **決定論的部分（ドリフト）**  
   → 長期トレンドを表すパラメータ $\mu$

2. **確率的部分（ランダム変動）**  
   → ボラティリティを表すパラメータ $\sigma$

標準偏差と分散の関係：

$ \sigma = \sqrt{\text{Var}} $

---

## 5. ウィーナー過程（Wiener Process）の性質

ウィーナー過程 $W_t$ の増分は次の性質を持つ：

$ W_{t+\Delta t} - W_t \sim \mathcal{N}(0, \Delta t) $

- 増分は独立  
- 増分は正規分布  
- サンプルパスは連続  

---

## 6. 幾何ブラウン運動（Geometric Brownian Motion: GBM）

金融工学で最も重要なSDE（確率微分方程式）はこれ：

$ dS_t = \mu S_t \, dt + \sigma S_t \, dW_t $

意味：

- $dS_t$：株価の微小変化  
- $\mu S_t dt$：決定論的トレンド  
- $\sigma S_t dW_t$：確率的変動（ボラティリティ）  
- $dW_t$：正規分布 $\mathcal{N}(0, dt)$ に従う増分

### 株価が大きいほど変動が大きくなる理由

式の通り：

$ \text{変動幅} \propto S_t $

→ 株価が成長すると変動が自然に大きくなる（実データと一致）。

---

## 7. 現代金融モデルへの応用

この

$ dS_t = \mu S_t dt + \sigma S_t dW_t $

は現代金融工学の中心であり：

- ブラック＝ショールズモデル  
- 金利モデル（Vasicek、CIR など）  
- デリバティブ価格付け  

といった多くの理論の基盤。

---

# まとめ

- リターンは正規分布  
- 株価は負の値を取れないので **対数正規分布**  
- 長期トレンドは $\mu$、短期変動は $\sigma$  
- ノイズ部分はウィーナー過程に従う  
- 株価変動の標準モデル：  
  $ dS_t = \mu S_t dt + \sigma S_t dW_t $



#11-4 ウィーナー過程とランダムウォークのシミュレーション

この資料は、講義で説明されていた「ウィーナー過程を Python でシミュレーションする話」を日本語で整理しつつ、実際に動かせるコードをオブジェクト指向（クラスベース）形式でまとめたものです。

---

## 1. 前提：リターンとランダム性

株価のリターン（日次リターン）は次の式で定義されます。

$ \text{Return} = \frac{\text{今日の株価} - \text{昨日の株価}}{\text{昨日の株価}} $

実データからリターンのヒストグラムを作ると次の特徴があります。

- 平均はほぼ 0  
- 正規分布に非常に近い形  

→ よって **リターンは正規分布に従う確率変数として扱える** と考えられます。

---

## 2. 株価そのものではなく「過程」をモデル化する理由

株価の動きを連続時間で変化する確率過程として扱うため、次の概念が登場します。

- ウィーナー過程（ブラウン運動）  
- ランダムウォーク  

ウィーナー過程 $W(t)$ の性質：

- **独立増分**  
- **正規増分**（平均 0、分散 $\Delta t$）  

式で表すと：

$ W(t + \Delta t) - W(t) \sim N(0, \Delta t) $

---

## 3. 数値シミュレーションの基本アイデア

連続時間の確率過程をコンピュータで扱うため、小さい時間刻み $dt$ を使って近似します。

手順：

1. 標準正規乱数 $Z \sim N(0,1)$ を生成  
2. 増分 $dW = \sqrt{dt} \times Z$ を計算  
3. 初期値 $W(0)=0$ とし、累積和で $W(t)$ を構築  

---

## 4. Python実装：ウィーナー過程クラス

以下に示すクラス **WienerProcess** は、ウィーナー過程の生成と可視化を扱います。

### 特徴
- 時間刻み $dt$ とステップ数 $N$ を設定  
- `generate_path`：パスを生成  
- `plot_path`：結果をプロット  

---

### コード：WienerProcess クラス

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

class WienerProcess:
    def __init__(self, dt=0.1, N=1000):
        """
        初期化
        dt : 時間刻み
        N  : ステップ数
        """
        self.dt = dt
        self.N = N
        self.t = None
        self.W = None

    def generate_path(self):
        """
        ウィーナー過程 W(t) を生成して内部に保持する
        """
        # 時間軸
        self.t = np.linspace(0, self.N * self.dt, self.N + 1)

        # 標準正規乱数 Z ~ N(0, 1)
        Z = npr.normal(0, 1, self.N)

        # 増分 dW = sqrt(dt) * Z
        dW = np.sqrt(self.dt) * Z

        # 初期値 0 の配列
        self.W = np.zeros(self.N + 1)

        # 累積和で W を構築
        self.W[1:] = np.cumsum(dW)

        return self.t, self.W

    def plot_path(self):
        """
        生成されたパスをプロットする
        """
        if self.W is None:
            self.generate_path()

        plt.figure(figsize=(10, 5))
        plt.plot(self.t, self.W, label="Wiener Process")
        plt.xlabel("Time")
        plt.ylabel("W(t)")
        plt.title("Wiener Process Simulation")
        plt.grid(True)
        plt.legend()
        plt.show()

# メイン処理
if __name__ == "__main__":
    process = WienerProcess(dt=0.1, N=1000)
    process.plot_path()
```

---

## 5. 株価モデルとしての評価

ウィーナー過程の特徴（欠点）

- 値が正にも負にもなる  
- 長期的トレンドが 0  
- ランダムなギザギザした動き  

一方、実際の株価の特徴は：

- 負にならない（最小値は 0）  
- 長期的に上昇トレンドをもつ  
- 株価が高いほど変動（ボラティリティ）も大きくなる  

→ よって、**ウィーナー過程そのままでは株価のモデルとして不適。**

---

## 6. 幾何ブラウン運動（GBM）への接続

株価のモデルには、次の確率微分方程式（SDE）が一般的に使われます。

$ dS(t) = \mu S(t)\,dt + \sigma S(t)\,dW(t) $

### 特徴

- 決定論的トレンド $\mu$（期待リターン）  
- ランダム変動 $\sigma$（ボラティリティ）  
- 解が指数関数形 → **負値にならない**  
- 現代金融工学の標準モデル  

---

## 7. まとめ

- リターンは正規分布として扱える  
- ウィーナー過程は最も基本的な確率過程  
- numpy の累積和で簡単にシミュレーション可能  
- しかし株価を表すには不十分（負値・トレンドがない）  
- そのため、**幾何ブラウン運動（GBM）** が株価モデルとして使われる  



````python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

class WienerProcess:
    def __init__(self, dt=0.1, N=1000):
        """
        初期化
        dt : 時間刻み
        N  : ステップ数
        """
        self.dt = dt
        self.N = N
        self.t = None
        self.W = None

    def generate_path(self):
        """
        ウィーナー過程 W(t) を生成して内部に保持する
        """
        # 時間軸
        self.t = np.linspace(0, self.N * self.dt, self.N + 1)

        # 標準正規乱数 Z ~ N(0, 1)
        Z = npr.normal(0, 1, self.N)

        # 増分 dW = sqrt(dt) * Z
        dW = np.sqrt(self.dt) * Z

        # 初期値 0 の配列
        self.W = np.zeros(self.N + 1)

        # 累積和で W を構築
        self.W[1:] = np.cumsum(dW)

        return self.t, self.W

    def plot_path(self):
        """
        生成されたパスをプロットする
        """
        if self.W is None:
            self.generate_path()

        plt.figure(figsize=(10, 5))
        plt.plot(self.t, self.W, label="Wiener Process")
        plt.xlabel("Time")
        plt.ylabel("W(t)")
        plt.title("Wiener Process Simulation")
        plt.grid(True)
        plt.legend()
        plt.show()

# メイン処理
if __name__ == "__main__":
    process = WienerProcess(dt=0.1, N=1000)
    process.plot_path()
````

#11-5 伊藤計算と確率変数の微分

この資料では、講義で説明されていた  
**「確率微分方程式（SDE）を解くための伊藤計算（Itô calculus）」**  
について、内容を整理したものです。

---

## 1. 普通の関数（確定変数）の微分

まず、確定変数 $x$ をもつ通常の関数を考えます。

例：  
$f(x) = x^2$

この微分は：

$\frac{df}{dx} = 2x$

小さな変化 $dx$ に対する変化は、

$df = 2x\,dx$

これは普通の微分の公式であり、  
関数が確定的で、$x$ がランダムでない場合に成り立つ。

---

## 2. 確率変数（ウィーナー過程）を含む場合はどうなるか？

今度は $x$ が **確率変数（例：ウィーナー過程 $W_t$）** の場合を考える。

ウィーナー過程の性質：

- $dW$ は平均 0、分散 $dt$ の正規分布  
- つまり  
  $dW \sim N(0, dt)$

ここで問題：

> 普通の微分法則 $df = f'(x)\,dx$ はそのまま使えるか？

→ **使えない。**

理由：  
$W_t$ は確率変数であり、$dW$ は無限小ではあるがランダムで、  
しかも二乗すると $dW^2 = dt$ という非常に特殊な性質を持つため。

---

## 3. 伊藤の補題（Itô's Lemma）

確率変数を含む関数 $f(x)$ の変化は次で与えられる：

$df = f'(x)\,dx + \frac{1}{2} f''(x)\,(dx)^2$

ここで、ウィーナー過程の性質

- $dW$ の期待値は 0
- $(dW)^2 = dt$

を使う。

ただの微分との違い：

- **2次の微小項 $(dx)^2$ が消えない**
- むしろ $(dW)^2 = dt$ となって残る

これが確率微分の最大のポイント。

---

## 4. Taylor 展開からの導出（直感）

関数 $f(x)$ を $x + dx$ で評価すると：

$f(x + dx) = f(x)  
\quad + f'(x)\,dx  
\quad + \frac{1}{2} f''(x)\,(dx)^2  
\quad + \cdots$

通常の微分では $(dx)^2$ は無視してよい小ささだが、  
確率変数では **無視すると間違いになる**。

なぜか？

- $dx = dW$ の場合  
- $dW \sim \sqrt{dt}$ の大きさ  
- $(dW)^2 = dt$ は無視できない

---

## 5. 実際に計算してみる：$f(x)=x^2$

### 関数
$f(x) = x^2$

### 1次微分
$f'(x) = 2x$

### 2次微分
$f''(x) = 2$

### 伊藤の補題を適用

$df = 2x\,dW + \frac{1}{2}\cdot 2 \cdot (dW)^2$

$dW^2 = dt$ を使って簡略化：

$df = 2x\,dW + dt$

---

## 6. 普通の微分との比較

確定変数の場合：

$df = 2x\,dx$

確率変数の場合（伊藤の補題）：

$df = 2x\,dW + dt$

**→ “+ dt” が追加されるのが決定的な違い。**

この追加項が後の

- 幾何ブラウン運動（GBM）
- Black–Scholes 方程式の導出
- 金融工学の確率微分方程式

すべてに現れる。

---

## 7. まとめ

- 普通の微分では 1次の微小項だけが残る  
- ウィーナー過程では $(dW)^2 = dt$ が残るため 2次項が重要  
- これを体系化したのが **伊藤の補題（Itô's Lemma）**  
- 金融工学では必須の道具  
- 次のステップは **幾何ブラウン運動（GBM）に伊藤の補題を適用すること**





#11-6 高次元伊藤（Itô）計算とブラック–ショールズ方程式の導入

この資料では、講義で説明されていた  
**「なぜ確率微分方程式（SDE）と伊藤計算が必要なのか」**  
を整理しつつ、ブラック–ショールズ式につながる *多変数の伊藤補題* をまとめる。

---

## 1. なぜ確率微分方程式（SDE）が必要なのか？

オプション価格 $V(S,t)$ を計算したいとき：

- 満期（$t = T$）ではオプションの価値は明確  
  例：コールなら $V = \max(S_T - K, 0)$

- しかし知りたいのは **満期前 $t < T$ のオプション価格**  
  → この時点では株価 $S_t$ が確率的に変動するため、  
     オプション価格 $V(S,t)$ も確率的に変動する

よって、株価と同じく **確率過程としてオプション価格を扱う必要がある**。

そのために **確率微分方程式（SDE）** と  
**伊藤計算（Itô calculus）** が必要になる。

---

## 2. 株価のモデル：SDE（一次元）

株価の変化は次の形でモデル化されたとする：

$ dS = a\,dt + b\,dW $

- $a = \mu S$（ドリフト：期待成長率）
- $b = \sigma S$（ボラティリティ）

ただし今回は記号を簡略化して $a, b$ をそのまま使う。

---

## 3. オプション価格は **2 変数の確率関数** になる

オプション価格 $V$ は次を変数に持つ：

- 株価：$S$（確率変数）
- 時間：$t$（確定変数）

つまり：

$ V = V(S, t) $

変化量は：

$ dV = V_S\,dS + V_t\,dt + \frac{1}{2}V_{SS}(dS)^2 $

多変数の伊藤の補題（2D Itô's Lemma）である。

---

## 4. Taylor 展開からの導出のイメージ

$V(S,t)$ を $(S + dS, t + dt)$ に広げると：

$ V(S+dS, t+dt) =  
V + V_S dS + V_t dt + \frac{1}{2}V_{SS}(dS)^2 + \cdots $

通常の微分なら $(dS)^2$ は無視して良いが、  
確率変数では **無視すると間違い**。

理由：  
$dS = a\,dt + b\,dW$ なので：

$(dS)^2 = (a\,dt + b\,dW)^2  
\approx b^2(dW)^2 = b^2\,dt$

ここでウィーナー過程の性質：

$(dW)^2 = dt$

が重要。

---

## 5. 必要な項だけ残すと…

$dS = a\,dt + b\,dW$ を使って：

$ (dS)^2 \approx b^2 dt $

これを Itô の式に代入：

$ dV  
= V_t dt  
+ V_S (a\,dt + b\,dW)  
+ \frac{1}{2} V_{SS} b^2 dt $

まとめると：

$ dV  
= \left( V_t + a V_S + \frac{1}{2}b^2 V_{SS} \right) dt  
+ b V_S\, dW $

これは **オプション価格の確率微分方程式** であり、  
後にブラック–ショールズ PDE に変形される。

---

## 6. なぜこれが重要なのか？

株価 $S$ が確率過程である以上：

- $S$ 依存のオプション価格 $V(S,t)$ も確率過程になる  
- よって $V$ の変化量 $dV$ を正しく扱うには  
  **高次元の伊藤補題** が不可欠

ブラック–ショールズの導出では：

- この式をリスクヘッジ資産に組み合わせることで  
- $dW$ の項を消し、リスクのないポートフォリオを作り  
- PDE（偏微分方程式）へ変形する

という流れになる。

---

## 7. まとめ

- 株価が確率過程 → オプション価格も確率過程  
- よって 2 変数の伊藤補題が必要  
- $dV$ には 1次項（$V_S$, $V_t$）だけでなく  
  **2次項の $\frac{1}{2}V_{SS}(dS)^2$ が必須**
- $(dS)^2 = b^2 dt$ という性質に注意  
- これが後のブラック–ショールズ方程式の核心部分



#11-7 幾何ブラウン運動（GBM）の確率微分方程式を伊藤の補題で解く

この資料では、講義で説明されていた  
**「幾何ブラウン運動（GBM）の SDE を伊藤の補題で解く手順」**  
をわかりやすく整理したものです。

---

## 1. 幾何ブラウン運動（GBM）の定義

株価 $S(t)$ の変化は次の確率微分方程式（SDE）で与えられる：

$ dS = \mu S\,dt + \sigma S\,dW $

- $\mu$：ドリフト（平均成長率）
- $\sigma$：ボラティリティ
- $W$：ウィーナー過程（ブラウン運動）

この式を **解いて明示的な形の $S(t)$** を得るのが今回の目的。

---

## 2. まず式を「対数微分の形」にする

両辺を $S$ で割る：

$ \frac{dS}{S} = \mu\,dt + \sigma\,dW $

これは「対数微分」と同じ形なので：

$ d(\ln S)$ を伊藤の補題で計算すれば良い。

よって、次の関数を定義する：

- $f(S) = \ln S$

---

## 3. 伊藤の補題（Itô's Lemma）を適用

伊藤の補題（1変数版）は：

$ df = f'(S)\,dS + \frac{1}{2} f''(S)\,(dS)^2 $

ここで：

### 1次微分
$f'(S) = \frac{1}{S}$

### 2次微分
$f''(S) = -\frac{1}{S^2}$

---

## 4. $(dS)^2$ を計算する

$ dS = \mu S\,dt + \sigma S\,dW $

の両辺を二乗する：

$(dS)^2  
= (\mu S\,dt + \sigma S\,dW)^2  
\approx \sigma^2 S^2 (dW)^2  
= \sigma^2 S^2 dt $

（理由：$dt^2$ や $dt\,dW$ は無視でき、$(dW)^2 = dt$ のみ残る）

---

## 5. 伊藤の補題に代入

$df = f'(S)\,dS + \frac{1}{2}f''(S)(dS)^2$

代入：

$df  
= \frac{1}{S}(\mu S\,dt + \sigma S\,dW)  
\quad + \frac{1}{2}\left(-\frac{1}{S^2}\right)(\sigma^2 S^2 dt)$

簡略化：

$df  
= \mu\,dt + \sigma\,dW  
\quad - \frac{1}{2}\sigma^2 dt$

よって：

$ d(\ln S)  
= \left(\mu - \frac{1}{2}\sigma^2\right) dt + \sigma\, dW $

---

## 6. 両辺を積分して明示解を得る

0〜t まで積分：

$\ln S(t) - \ln S(0)  
= \left(\mu - \frac{1}{2}\sigma^2\right)t + \sigma W(t)$

両辺に指数をつけて $S(t)$ を求める：

$ S(t)  
= S(0)\,\exp\left[\left(\mu - \frac{1}{2}\sigma^2\right)t + \sigma W(t)\right] $

---

## 7. 幾何ブラウン運動の解の特徴

- **常に正の値をとる**（指数関数形のため）
- **平均成長率 $\mu$ に従って長期的に増えていく**
- **ボラティリティ $\sigma$ が大きいほど変動も大きい**
- **株価の現実的特徴を非常によく表す**

→ この理由で **GBM は現代金融工学の標準的な株価モデル** になっている。

---

## 8. まとめ

- SDE $dS=\mu S dt + \sigma S dW$ を伊藤の補題で解いた  
- 対数変換 $f(S)=\ln S$ を使うのがポイント  
- $(dW)^2 = dt$ を利用して 2次項が残る  
- 最終解は：

$ S(t) = S(0)\exp\big[(\mu - \tfrac{1}{2}\sigma^2)t + \sigma W(t)\big] $

- これが次の講義でシミュレーションする **GBM の明示解** となる



#11-8 幾何ブラウン運動（GBM）のPython実装

この資料では、講義で説明されていた  
**「幾何ブラウン運動（Geometric Brownian Motion: GBM）のシミュレーション」**  
を、数式の復習から Python 実装まで整理してまとめる。

---

## 1. 幾何ブラウン運動の復習

前の講義で学んだ GBM の確率微分方程式（SDE）は：

$ dS = \mu S\,dt + \sigma S\,dW $

この式の解（明示解）は：

$ S(t) = S_0 \exp\left[ \left(\mu - \frac{1}{2}\sigma^2\right)t + \sigma W(t) \right] $

したがって、Python では

- $W(t)$ を Brownian motion（累積和）
- その結果を上式に代入

することで株価パスを生成できる。

---

## 2. GBM シミュレーションのアルゴリズム

1. 時間刻み  
   $dt = T/N$ を計算する  
2. 時間軸  
   `np.linspace(0, T, N+1)`  
3. 標準正規乱数を生成  
   $Z \sim N(0,1)$  
4. $W(t)$ を累積和で生成  
   $W(t) = \sqrt{dt} \sum Z$  
5. 明示解  
   $S(t) = S_0 \exp[(\mu - \tfrac12\sigma^2)t + \sigma W(t)]$

---

## 3. Python コード：幾何ブラウン運動（GBM）の実装

以下は講義内容に基づいた **シンプルで本格的な実装**。

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_gbm(S0, T, N, mu, sigma):
    """
    幾何ブラウン運動 (GBM) のシミュレーション
    S0   : 初期株価
    T    : シミュレーション期間（年）
    N    : ステップ数
    mu   : ドリフト（平均成長率）
    sigma: ボラティリティ
    """
    # 1. 時間刻み
    dt = T / N

    # 2. 時間軸
    t = np.linspace(0, T, N + 1)

    # 3. 標準正規乱数
    Z = np.random.normal(0, 1, N)

    # 4. Brownian motion
    W = np.zeros(N + 1)
    W[1:] = np.cumsum(np.sqrt(dt) * Z)

    # 5. GBM の明示解
    S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

    return t, S

def plot_gbm(t, S):
    plt.figure(figsize=(10, 5))
    plt.plot(t, S, label="Geometric Brownian Motion")
    plt.xlabel("Time")
    plt.ylabel("Stock Price")
    plt.title("GBM Simulation")
    plt.grid(True)
    plt.legend()
    plt.show()

# メイン処理例
if __name__ == "__main__":
    S0 = 10       # 初期株価
    T = 2         # 2年間
    N = 1000      # 1000ステップ
    mu = 0.1      # 平均成長率（10%）
    sigma = 0.3   # ボラティリティ

    t, S = simulate_gbm(S0, T, N, mu, sigma)
    plot_gbm(t, S)
```

---

## 4. 出力される株価軌道の特徴

シミュレーション結果を確認すると：

- **上昇トレンド（μによる）**
- **ランダム変動（σによる）**
- **価格は常に 0 より大きい** → 株価の性質に合致  
- **σ を大きくすると変動が大きくなる**
- **S0 を変えると開始点が変わる**

たとえば：

- `sigma = 0.05` → 滑らかで安定  
- `sigma = 0.5` → 激しく上下に揺れる  

---

## 5. このモデルが株価に適している理由

幾何ブラウン運動 GBM は株価の性質を自然に再現する：

- **負の値を取らない（下限が 0）**
- **長期的に増加する傾向（経済成長）**
- **価格が大きいと変動幅も大きくなる（対数正規性）**
- **金融工学で最も広く使われる標準モデル**

（Black–Scholes 価格公式も GBM を前提に導かれる）

---

## 6. まとめ

- GBM の SDE：$dS = \mu S dt + \sigma S dW$  
- 伊藤の補題で解くと明示式が得られる  
- Brownian motion を累積和で生成し、指数に代入するだけでシミュレーション可能  
- Python では NumPy を使って効率的に実装できる  
- 株価の現実的特徴（対数正規性・非負性）を自然に満たす  





````python
import numpy as np
import matplotlib.pyplot as plt

def simulate_gbm(S0, T, N, mu, sigma):
    """
    幾何ブラウン運動 (GBM) のシミュレーション
    S0   : 初期株価
    T    : シミュレーション期間（年）
    N    : ステップ数
    mu   : ドリフト（平均成長率）
    sigma: ボラティリティ
    """
    # 1. 時間刻み
    dt = T / N

    # 2. 時間軸
    t = np.linspace(0, T, N + 1)

    # 3. 標準正規乱数
    Z = np.random.normal(0, 1, N)

    # 4. Brownian motion
    W = np.zeros(N + 1)
    W[1:] = np.cumsum(np.sqrt(dt) * Z)

    # 5. GBM の明示解
    S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

    return t, S

def plot_gbm(t, S):
    plt.figure(figsize=(10, 5))
    plt.plot(t, S, label="Geometric Brownian Motion")
    plt.xlabel("Time")
    plt.ylabel("Stock Price")
    plt.title("GBM Simulation")
    plt.grid(True)
    plt.legend()
    plt.show()

# メイン処理例
if __name__ == "__main__":
    S0 = 10       # 初期株価
    T = 2         # 2年間
    N = 1000      # 1000ステップ
    mu = 0.1      # 平均成長率（10%）
    sigma = 0.3   # ボラティリティ

    t, S = simulate_gbm(S0, T, N, mu, sigma)
    plot_gbm(t, S)
````

#12 ブラックショールズモデル

#12-1 ブラック＝ショールズモデル入門：リスク中立戦略とデルタヘッジ

この章では、これまで学んだ  
- ブラウン運動  
- 幾何ブラウン運動  
- 伊藤の補題（Ito’s Lemma）  
- 確率微分方程式（SDE）  

といった準備をもとに、**ブラック＝ショールズ（Black-Scholes）モデル** の直前までを整理する。

Black–Scholes（Black & Scholes, 1973）は、金融工学革命の中心にある理論であり、  
「オプション価格を満期前にどう評価するか？」という問題を解いた。

---

## 1. ブラック＝ショールズモデルの目的

オプションは「将来のある時点（満期 T）での価値」は明確に分かる。

- コール：$\max(S_T - K, 0)$  
- プット：$\max(K - S_T, 0)$  

しかし知りたいのは **T より前の時点（t < T）での公正価値**。

Black–Scholes の目的は：

> **満期前のオプション価格を理論的に決めること**

---

## 2. 驚くべき発見：「リスク資産を組み合わせるとリスクが消える」

一般的な金融理論（モダンポートフォリオ理論、CAPM）は：

- 株価はランダムに動く  
- リスクは不可避  
- 分散投資で「銘柄固有のリスク」は消せるが  
  **市場全体のリスク（systematic risk）は消せない**

という前提だった。

しかし Black-Scholes モデルでは次を示した：

> **リスク資産を特定の割合で組み合わせると、  
> 市場リスクすら完全に消える（＝無リスク化）**

これは非常に革命的だった。

---

## 3. オプション価格が依存する要素

オプション価格 $V(S, t)$ は複数の要因に依存する。

### 変数（状態量）
- $S$: 基礎資産（株価）
- $t$: 時間

### パラメータ（モデル特性）
- $\mu$: 株の期待成長率  
- $\sigma$: ボラティリティ  
- $K$: 権利行使価格（ストライク）  
- $T$: 満期  
- $r$: 無リスク金利  

---

## 4. コールオプションとプットオプションの性質

### コール（Call）
- 権利：未来に **買う** 権利
- 価値：$S$ が上がるほど上昇  
  → **基礎資産と正の相関**

### プット（Put）
- 権利：未来に **売る** 権利
- 価値：$S$ が下がるほど上昇  
  → **基礎資産と負の相関**

この「相関」がデルタヘッジの根幹となる。

---

## 5. リスクを消すための特別なポートフォリオ

ブラック＝ショールズが考えたのは以下の組み合わせ：

\[
\Pi = V - \Delta S
\]

つまり、

- **オプションをロング（買い）**
- **株を Δ だけショート（売り）**

というポートフォリオ。

重要な点：

- コールでは $V$ と $S$ の動きは **同じ方向**（正相関）  
- 適切な Δ を選ぶと、  
  **オプションのランダムな変動を株の反対売買で完全に打ち消せる**

これを **デルタヘッジ（Delta Hedging）** という。

結果：

> **ポートフォリオが無リスク化される → 無リスク金利 r で成長するに決まっている**

ここからブラック＝ショールズ偏微分方程式が導かれる。

---

## 6.まとめ

- ブラック＝ショールズモデルの目的は「満期前のオプション価格」を決定すること  
- ランダムな価格変動を持つ株とオプションを組み合わせると  
  **リスクを完全に除去できる（デルタヘッジ）**  
- これは市場リスクすら消えるため、非常に革命的な発見  
- 次章では、このデルタヘッジを数学的に扱い、  
  **ブラック＝ショールズ偏微分方程式（PDE）** を導く  



#12-2 ブラック＝ショールズモデル：デルタヘッジによるリスク消去の仕組み

前章では、ブラック＝ショールズの発想の中心となる  
**「オプションと株を組み合わせるとリスクを消せる」**  
という驚くべきアイデアの背景を見た。

本章では、その具体的な数学的メカニズム、特に  
**ポートフォリオの変化量とデルタヘッジの導出**  
を詳しく解説する。

---

## 1. 前提：株価のモデル（幾何ブラウン運動）

基礎資産 $S(t)$ は次の確率微分方程式（SDE）に従う：

$$
dS = \mu S\, dt + \sigma S\, dW
$$

- $\mu S dt$：決定論的な平均成長  
- $\sigma S dW$：ランダム変動  
- $dW$：Wiener 過程の微小変化  

---

## 2. 特別なポートフォリオの構築

ブラック＝ショールズの核心は次の組み合わせ：

$$
\Pi = V - \Delta S
$$

- $V(S,t)$：オプション価格  
- $S$：基礎資産  
- $\Delta$：ショートする株数  

**オプションをロングし、株を $\Delta$ だけショートする。**

目標：

> **このポートフォリオのリスク（確率的成分）を完全に消すこと**

---

## 3. ポートフォリオの変化：$d\Pi$

ポートフォリオの微小変化は：

$$
d\Pi = dV - \Delta\, dS
$$

ここで、$V$ も $S$ も確率変数 → $dV$ を求めるために **伊藤の補題** が必要。

---

## 4. 伊藤の補題（二変数版）による $dV$

オプション価格 $V(S,t)$ に伊藤の補題を適用すると：

$$
dV = V_t dt + V_S dS + \frac{1}{2} V_{SS} (dS)^2
$$

---

## 5. $(dS)^2$ の計算

株価の SDE：

$$
dS = \mu S dt + \sigma S dW
$$

これを二乗して小さい項を無視すると：

- $(dt)^2$ → 無視  
- $dt \cdot dW$ → 無視  
- $(dW)^2 = dt$

よって、

$$
(dS)^2 \approx \sigma^2 S^2 dt
$$

---

## 6. $dV$ を具体的な形に展開

$$
dV
= \left( V_t + \mu S V_S + \frac{1}{2}\sigma^2 S^2 V_{SS} \right) dt
\;+\;
\sigma S V_S\, dW
$$

右端の $dW$ が **確率的ゆらぎ（リスク）**。

---

## 7. ポートフォリオの変化 $d\Pi$

$$
d\Pi = dV - \Delta dS
$$

$ dS = \mu S dt + \sigma S dW $ を入れると：

$$
d\Pi =
\left( V_t + \mu S V_S + \frac12 \sigma^2 S^2 V_{SS} - \Delta \mu S \right) dt
+
\left( \sigma S V_S - \Delta \sigma S \right) dW
$$

---

## 8. 確率項の消去：デルタヘッジ

確率的成分（$dW$ の係数）は：

$$
\sigma S V_S - \Delta \sigma S
$$

これを 0 にすれば **完全にリスクが消える**。

$$
\sigma S V_S - \Delta \sigma S = 0
$$

したがって：

$$
\Delta = V_S
$$

これが **デルタヘッジ**。

---

## 9. デルタヘッジの意味

- $\Delta = V_S$：オプション価格の株価に対する感応度  
- これだけ株をショートすると、オプションの確率変動と正確に相殺される  

つまり：

> **2つのリスク資産を組み合わせることで、リスクゼロの資産を「人工的に」作れる**

これがブラック＝ショールズの革命。

---

## 10. デルタは「動的に」変化する

$\Delta = V_S$ は時間とともに必ず変わる。

- $S$ が変わる  
- $t$ が進む  
- $\sigma$ や他の条件も影響  

そのため：

> **ヘッジは常に調整されねばならない → 動的デルタヘッジ**

---

## 11. 次章へ

デルタヘッジによりポートフォリオは無リスクになる。  
→ 無リスク資産なら期待収益率は **必ず無リスク金利 $r$**。

この性質を用いて、

- ブラック＝ショールズ偏微分方程式（PDE）が導かれる  
- さらに閉形式解（Black–Scholes 公式）へつながる  

次章ではその PDE を実際に導出する。



#12-3 ブラック＝ショールズ方程式（Black–Scholes PDE）への導出  
## ノーアービトラージ原理から導く決定論的ポートフォリオの変化

前章では、次の重要な事実を確認した：

- オプションと株を適切に組み合わせたポートフォリオ  
- 伊藤の補題による $dV$ の展開  
- デルタ $\Delta = V_S$ を選ぶことで確率項 $dW$ が完全に消える（デルタヘッジ）

その結果：

> **ポートフォリオの変化 $d\Pi$ は “完全に決定論的” になる（リスクゼロ）。**

この章では、その決定論的変化を **ノーアービトラージ原理** と結びつけて、  
ブラック＝ショールズ偏微分方程式（PDE）を導く。

---

## 1. リスクを消した結果：ポートフォリオは完全に「確定的」になる

デルタヘッジ後のポートフォリオ：

$$
\Pi = V - \Delta S
$$

デルタ $\Delta = V_S$ を選ぶと、

$$
d\Pi \text{ から確率項 } dW \text{ が完全に消える}
$$

つまり：

> **このポートフォリオは “リスクゼロの資産” として振る舞う。**

---

## 2. ノーアービトラージ原理（No-Arbitrage Principle）

金融の基本原則：

> **リスクゼロの資産は、常にリスクフリー金利 $r$ で成長しなければならない。**

理由：

- リスクなしで市場平均以上の収益が得られたら  
  → 誰でも借金して投資し無限に儲けられてしまう  
- これは **アービトラージ（無リスク裁定）** であり  
  → 市場はそれを許さない（瞬時に解消される）

したがって、

$$
d\Pi = r \Pi\, dt
$$

でなければならない。

---

## 3. ポートフォリオの決定論的変化

前章の結果より、デルタヘッジ後の $d\Pi$ の決定論的成分は：

$$
d\Pi
=
\left( V_t
+ \frac12 \sigma^2 S^2 V_{SS}
+ \mu S V_S
- \Delta \mu S
\right) dt
$$

デルタ $\Delta = V_S$ を代入すると、$\mu$ の項が消える：

$$
d\Pi
=
\left(
V_t + \frac12 \sigma^2 S^2 V_{SS}
\right) dt
$$

---

## 4. ノーアービトラージ条件と結びつける

ポートフォリオ $\Pi = V - V_S S$ を使い、

$$
d\Pi = r \Pi\, dt
$$

すなわち：

$$
V_t + \frac12 \sigma^2 S^2 V_{SS}
=
r (V - S V_S)
$$

---

## 5. ブラック＝ショールズ偏微分方程式（Black–Scholes PDE）

式を整理すると：

$$
V_t + \frac12 \sigma^2 S^2 V_{SS}
+ r S V_S - r V = 0
$$

これは金融工学で最も重要な PDE の一つ：

> **ブラック＝ショールズ方程式（Black–Scholes PDE）**

---

## 6. なぜこの方程式は重要なのか？

- この PDE の解がヨーロピアン・オプション価格  
- 数学的に厳密に導かれたモデル  
- 市場が「裁定機会なし」であるための **唯一の価格** を与える  
- これに境界条件（満期でのペイオフ）を入れて解くと  
  → 有名なブラック＝ショールズ closed-form solution が得られる

---

## 7. 本章のまとめ

- デルタヘッジによりポートフォリオは確率項が消え、リスクゼロになる  
- ノーアービトラージ原理より、その成長率は必ず $r$  
- この条件を満たすことが、ブラック＝ショールズ PDE を強制する  
- PDE の解がオプション価格そのものになる  

---

## 8. 次章へ

次章ではこの PDE を実際に解き、  
**ブラック＝ショールズ公式（closed-form solution）**  
を導いていく。



#12-4 ブラック＝ショールズ公式（Black–Scholes Closed-Form Solution）
## PDE の解としてのコール・プット価格公式

前章では、デルタヘッジとノーアービトラージ原理から  
**ブラック＝ショールズ方程式（PDE）** を導いた。

本章では、その PDE の「解」がどのような形になるのかを説明する。

---

## 1. ブラック＝ショールズ方程式（おさらい）

オプション価格 $V(S,t)$ は次の PDE を満たす：

$$
V_t + \frac12 \sigma^2 S^2 V_{SS} + r S V_S - r V = 0
$$

これは数学的には **線形・パラボリック型 PDE**。

> 代表例：熱方程式（heat equation）、拡散方程式（diffusion equation）

金融数学においても同じ数学構造が出てくる。

---

## 2. PDE を解く必要は実務では「ない」

ブラック＝ショールズ方程式の解法は高度な数学を含む：

- 伊藤積分
- 変数変換  
- 熱方程式への写像
- ガウス積分

しかし、**最終結果（closed-form solution）は非常にシンプル**で、  
実務ではこれをそのまま使えばよい。

講義でも強調された通り：

> オプション価格の計算では、SDE や伊藤の補題を理解していなくても  
> **最終式さえ使えれば問題ない**。

---

## 3. 標準正規分布 $N(x)$ の定義

ブラック＝ショールズ公式では **標準正規分布の累積分布関数（CDF）** を使う：

$$
N(x) = \frac{1}{\sqrt{2\pi}}
\int_{-\infty}^{x} e^{-z^2 / 2} \, dz
$$

実装では NumPy や SciPy に含まれているので、自分で積分する必要はない。

---

## 4. 必要となるパラメータ $d_1$ と $d_2$

ブラック＝ショールズ公式では、まず次の二つを計算する：

### $d_1$ の定義

$$
d_1 =
\frac{
\ln\left(\frac{S}{K}\right)
+ \left(
r + \frac12 \sigma^2
\right)(T - t)
}{
\sigma \sqrt{T - t}
}
$$

### $d_2$ の定義

$$
d_2 = d_1 - \sigma \sqrt{T - t}
$$

意味：

- $\ln(S/K)$：現在の株価と行使価格の相対関係  
- $r$：無リスク金利  
- $\sigma$：ボラティリティ  
- $T-t$：満期までの時間  

---

## 5. コールオプション（Call）の価格

ブラック＝ショールズ公式：

$$
C = S\, N(d_1) - K e^{-r(T - t)} N(d_2)
$$

---

## 6. プットオプション（Put）の価格

同様に、プットの価格は：

$$
P = K e^{-r(T - t)} N(-d_2) - S\, N(-d_1)
$$

---

## 7. 公式さえあれば実装はシンプル

重要ポイント：

- PDE や SDE を知らなくても  
  **$d_1$、$d_2$、$N(\cdot)$ を計算するだけで良い**
- NumPy / SciPy に CDF がある  
- 計算式は短く、実務実装は非常に簡単

次章では、これらの公式を **Python（クラスベース）で実装**する。

---

## まとめ

- ブラック＝ショールズ PDE は高度だが、最終解は単純  
- 標準正規分布の累積分布関数 $N(x)$ が重要  
- $d_1$・$d_2$ を計算し、コール/プット公式に代入するだけ  
- 実務ではこの closed-form solution がそのまま使われる  





#12-5 オプション価格の感応度指標 ― Greeks（グリークス）

ブラック＝ショールズモデルを理解したら、次に重要となるのが  
**Greeks（グリークス）** と呼ばれる感応度指標である。

Greeks は、オプション価格が何に対してどれだけ敏感かを測る尺度であり、  
ヘッジング・リスク管理・ポートフォリオ最適化で中心的な役割を果たす。

---

## 1. Delta（デルタ）

### 定義  
デルタはオプション価格 $V$ の株価 $S$ に対する一次微分：

$$
\Delta = \frac{\partial V}{\partial S}
$$

意味：  
- **オプションの価格が、株価のわずかな変化に対してどれくらい変わるか**
- オプションの「株的感応度」

### 重要な性質  
ブラック＝ショールズのデルタヘッジで登場したように：

> $\Delta$ を株のショート量として用いると、リスクが完全に消える（$dW$ が消える）

すなわち：

- ポートフォリオ  
  $\Pi = V - \Delta S$  
  の確率項を 0 にできる  
- デルタヘッジは **リスクの完全除去** を目的とした戦略

---

## 2. Gamma（ガンマ）

### 定義  
ガンマはオプション価格の二次微分：

$$
\Gamma = \frac{\partial^2 V}{\partial S^2}
$$

意味：

- **デルタが株価の変化に対してどれだけ変わるか（デルタの感応度）**
- デルタの変化率（カーブの強さ）

### ガンマの役割  
- ガンマが大きいほど、デルタ値が急激に変動する  
- つまり **ヘッジの調整頻度が増える**
- デルタヘッジが「動的」でなければいけない理由

---

## 3. Theta（セータ）

### 定義  
オプション価格の時間に対する微分：

$$
\Theta = \frac{\partial V}{\partial t}
$$

意味：

- **時間の経過によって、オプション価格がどれだけ減少するか**
- 特にオプションは時間価値を持つため、多くの場合 $\Theta < 0$

別名：  
- **タイムディケイ（Time decay）**

---

## 4. Vega（ベガ）

### 定義  
ボラティリティ $\sigma$ に対する感応度：

$$
\text{Vega} = \frac{\partial V}{\partial \sigma}
$$

意味：

- **ボラティリティが変わるとオプション価格がどれだけ動くか**

ポイント：

- ボラティリティは市場予測で大きくブレる  
- そのため **ボラティリティリスクは非常に重要**

---

## 5. Rho（ロー）（補足）

金利 $r$ に対する感応度：

$$
\rho = \frac{\partial V}{\partial r}
$$

金利変化がオプション価格に与える影響を示す。

---

## 6. ブラック＝ショールズ PDE を Greeks で書き換える

ブラック＝ショールズ PDE：

$$
V_t + \frac12\sigma^2 S^2 V_{SS} + r S V_S - rV = 0
$$

Greeks を用いると：

- $V_t = \Theta$  
- $V_S = \Delta$  
- $V_{SS} = \Gamma$

これにより、

$$
\Theta + \frac12 \sigma^2 S^2 \Gamma + r S \Delta - rV = 0
$$

このように、PDE を「Greeks の組み合わせ」として表現できる。

---

## 7. Greeks のまとめ

| Greek | 数式 | 意味 |
|-------|------|------|
| $\Delta$ | $\partial V / \partial S$ | 株価への感応度（一次） |
| $\Gamma$ | $\partial^2 V / \partial S^2$ | デルタの感応度（二次） |
| $\Theta$ | $\partial V / \partial t$ | 時間経過の影響（タイムディケイ） |
| Vega | $\partial V / \partial \sigma$ | ボラティリティの影響 |
| $\rho$ | $\partial V / \partial r$ | 金利の影響 |

---




#12-6 Black–Scholes モデルで “どう儲けるか？”

ここまでの章では、ブラック＝ショールズ方程式の導出、解法、Python 実装、  
さらにモンテカルロ・シミュレーションまで扱ってきた。

本章では、より実践的なテーマとして：

> **Black–Scholes モデルを使ってどう利益を出すのか？**

について整理する。

---

## 1. Black–Scholes モデルの用途は「儲け」だけではない

まず重要なのは、すべての投資家が “儲ける目的” でこのモデルを使うわけではない。

多くのプロ投資家の主目的は **リスクの除去（ヘッジ）** である。

例：

- ポートフォリオ全体のリスク管理  
- デルタやガンマを毎日計算してヘッジ量を調整  
- 市場変動に対してポジションを中立にする

つまり、Black–Scholes モデルは **利益よりもリスク制御のために使われることが多い**。

---

## 2. Black–Scholes を使った「利益の取り方」は？

もちろん **儲けるために使う** ことも可能である。

### 2.1. 方法①：ミスプライス（誤価格）を見つける

市場で取引されているオプション価格 $P_\text{market}$ が  
Black–Scholes が導く理論価格 $P_\text{BS}$ と異なる場合、

$$
P_\text{market} \neq P_\text{BS}
$$

という **ミスプライス（誤価格）** が存在する。

#### 例：  
- 市場価格が理論価格より安い  
  → 価値より安いので **買い**（期待値プラス）  
- 市場価格が理論価格より高い  
  → 過大評価されているので **売り**（またはショート）

ミスプライスは常に発生しており、  
多くのトレーダーはブラック＝ショールズを用いて “適正価格との差” を監視している。

---

### 2.2. 方法②：ボラティリティの歪みを利用する（IV アービトラージ）

市場では、銘柄・満期・ストライクによって  
インプライド・ボラティリティ(IV) に歪みがある。

```
同じ銘柄なのに  
・コールとプットで IV が違う  
・満期の違いで IV が不自然に上下する  
```

こうした歪みを使うと：

- 安い IV のオプションを買う  
- 高い IV のオプションを売る  

などの戦略で利益を狙える。

---

### 2.3. 方法③：ブラック＝ショールズを基準にした統計的裁定取引

株式でやった「ペアトレード」や  
FX で学んだ「ベルマン–フォード法でのアービトラージ検出」と同様に、

オプション市場にも統計的・構造的な歪みが存在するため

- モデル vs 市場のズレ  
- 複数オプション間の価格関係の不整合  

などを利用して利益を狙うことができる。

---

## 3. どのように利益を確定させるのか？

### 3.1. 株式での利益方法との比較

株で儲ける方法は基本的に：

- **上がると思ったら買う** → 上がったら売る  
- **下がると思ったら売る** → 下がったら買い戻す  

だがオプションの場合は、

- **価格の “適正さ” を判断できる**  
- **ボラティリティの歪みも利用できる**

という点で株より多くの戦略が存在する。

---

## 4. Black–Scholes モデルでの「儲け」のまとめ

| 方法 | 説明 |
|------|------|
| ミスプライスの発見 | 市場価格と理論価格の差で裁定 |
| IV（ボラティリティ）の歪み利用 | 高いIVを売り、安いIVを買う |
| 統計的アービトラージ | 複数オプションの価格関係の歪み利用 |
| リスク除去（ヘッジ） | 利益というより損失回避が目的 |

---

## 5. 結論

Black–Scholes モデルは、  
**オプションの“正しい価値”を教えてくれる道具** であり、

- リスク管理（ヘッジ）
- 価格の歪み発見（アービトラージ）
- ボラティリティ取引

など、プログラミングと数学を活かした高度な投資方法につながる。




#12-7 Long-Term Capital Management（LTCM）と量的金融モデルの成功

この章では、1990年代に大成功し、同時に劇的な崩壊を経験した  
**LTCM（Long-Term Capital Management）** という伝説的ヘッジファンドについてまとめる。

---

## 1. LTCM とは？

LTCM は 1994 年に設立されたヘッジファンドで、メンバーの多くがアカデミック出身だった。

- Salomon Brothers 副会長：John Meriwether  
- ノーベル経済学賞受賞者：Myron Scholes、Robert C. Merton

彼らは高度な数量モデルを積極的に使用した：

- ペアトレード（Pairs Trading）
- Black–Scholes モデル
- 統計的裁定取引（Statistical Arbitrage）

### 驚異のリターン：年率 40%

銀行預金の金利（1–2%）と比較すると異常な高さ。  
米国債（T-Bill）でも 4–5%。  

**LTCM は 3年以上連続で 40% のリターンを実現** した。

---

## 2. なぜそれほど儲かったのか？

### → 答え：**市場リスクすらヘッジする戦略を構築したから**

現代ポートフォリオ理論（MPT）では：

- 分散投資で **非システマティックリスク**（企業固有リスク）は削減可能
- **システマティックリスク**（市場全体のリスク）は避けられない  
  → CAPM のベータがその尺度

LTCM が革新的だったのは：

> **市場リスク（システマティックリスク）すらヘッジしたこと**

つまり、

- 非システマティックリスク  
- システマティックリスク  

**両方のリスクをほぼゼロに近づけた。**

---

## 3. どうやって市場リスクまで消したのか？

### ➤ 市場中立戦略（Market Neutral Strategy）

LTCM の中心となった戦略は次の2つ：

1. **Black–Scholes モデルによるデルタヘッジ**
2. **ペアトレード（Pairs Trading）**

どちらも **長期ポジション（Long）と短期ポジション（Short）を同時に持つ**  
というのが特徴。

### 3.1 Black–Scholes のデルタヘッジ

オプション価格と基礎資産（株価）は強い相関を持つ。

- コールオプション → 正の相関  
- プットオプション → 負の相関

この相関を利用し、

> **ロング：オプション**  
> **ショート：基礎株式の一定量（デルタ分）**

というポートフォリオを作ると、  
ランダムな価格変動（リスク）を完全に打ち消すことができる。

### 3.2 ペアトレード（Pairs Trading）

類似企業（コカコーラ vs ペプシのような）の株価には  
**歴史的な強い相関**がある。

例：

- ペプシの株価だけ急上昇  
- コカコーラは変化なし  

→ このとき：

- **割高（上がりすぎ）＝ペプシをショート**
- **割安（出遅れ）＝コカコーラをロング**

相関が戻ると価格差が縮まり利益になる。

これも **市場リスクを打ち消す典型的な市場中立戦略**。

---

## 4. なぜ崩壊したのか？（1998年）

アジア通貨危機・ロシア金融危機によって市場構造が急変し、

- 相関が崩壊  
- 流動性が消滅  
- 超レバレッジ（30倍以上）が裏目に出る

結果として LTCM は破綻し、連邦準備制度（FRB）が救済に介入した。

---

## 5. まとめ：LTCM から学べること

### ✔ 成功の理由  
- 高度な数量モデル  
- 市場中立戦略を極限まで洗練  
- 市場リスクさえヘッジ可能と考えた

### ✔ 失敗の理由  
- 市場相関が永遠に続くと仮定

#13-1 ブラック＝ショールズモデル実装（Python）  
この章では、ブラック＝ショールズモデルの理論背景を踏まえつつ、  
**Python によるブラック＝ショールズ方程式の実装**を行う。

SciPy の正規分布関数を利用することで、数学的な導出の複雑さを意識せずとも  
シンプルに実装できることがポイント。

---

## 1. 必要ライブラリの読み込み

ブラック＝ショールズモデルの公式は、正規分布の累積分布関数（CDF）を使用するため、  
SciPy の `stats.norm.cdf` を使う。また NumPy の `log`, `exp`, `sqrt` も利用する。

```python
from scipy import stats
from numpy import log, exp, sqrt
```

---

## 2. D1, D2 の計算式

ブラック＝ショールズモデルでは、まず以下の 2 つの値を計算する：

- **$D_1$**  
  $D_1 = \frac{\ln(S/K) + (r + \frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}$

- **$D_2$**  
  $D_2 = D_1 - \sigma\sqrt{T}$

ここで  

- $S$ … 現在の株価  
- $K$ … 権利行使価格（ストライク）  
- $T$ … 残存期間（年）  
- $r$ … 無リスク金利  
- $\sigma$ … ボラティリティ  

---

## 3. コールオプション価格の実装

ブラック＝ショールズのコール価格公式：

$C = S\,N(D_1)\;-\;K e^{-rT} N(D_2)$

```python
def call_option_price(S, K, T, r, sigma):
    # D1
    d1 = (log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * sqrt(T))
    # D2
    d2 = d1 - sigma * sqrt(T)

    # コールオプション価格
    call = S * stats.norm.cdf(d1) - K * exp(-r * T) * stats.norm.cdf(d2)
    return call
```

---

## 4. プットオプション価格の実装

ブラック＝ショールズのプット価格公式：

$P = K e^{-rT} N(-D_2)\;-\;S\,N(-D_1)$

```python
def put_option_price(S, K, T, r, sigma):
    d1 = (log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * sqrt(T))
    d2 = d1 - sigma * sqrt(T)

    put = K * exp(-r * T) * stats.norm.cdf(-d2) - S * stats.norm.cdf(-d1)
    return put
```

---

## 5. メイン処理（実行例）

以下の例では、ブラック＝ショールズ公式が正しく機能しているかを確認する。

## 7. ブラック＝ショールズモデルが強力な理由

---

## 8. まとめ

- $D_1$、$D_2$ の計算が最初のステップ  
- SciPy により正規分布は自前実装不要  
- Python でコール/プット価格が簡潔に実装できる

以上がブラック＝ショールズモデルの完全な Python 実装。



````python
from scipy import stats
from numpy import log, exp, sqrt

def call_option_price(S, K, T, r, sigma):
    # D1
    d1 = (log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * sqrt(T))
    # D2
    d2 = d1 - sigma * sqrt(T)

    # コールオプション価格
    call = S * stats.norm.cdf(d1) - K * exp(-r * T) * stats.norm.cdf(d2)
    return call

def put_option_price(S, K, T, r, sigma):
    d1 = (log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * sqrt(T))
    d2 = d1 - sigma * sqrt(T)

    put = K * exp(-r * T) * stats.norm.cdf(-d2) - S * stats.norm.cdf(-d1)
    return put

if __name__ == "__main__":
    S = 100      # 現在の株価
    K = 100      # ストライク
    T = 1        # 残存期間（1年）
    r = 0.05     # 金利5%
    sigma = 0.20 # ボラティリティ20%

    call = call_option_price(S, K, T, r, sigma)
    put = put_option_price(S, K, T, r, sigma)

    print("D1 and D2 are calculated internally.")
    print(f"Black–Scholes Call Price = {call:.4f}")
    print(f"Black–Scholes Put Price  = {put:.4f}")
````

#13-2 モンテカルロシミュレーションの理論背景

この章では、ブラック–ショールズモデルをモンテカルロ法で数値的に解くための  
**理論背景（Monte Carlo Simulation）** を整理します。

---

## 1. モンテカルロ法とは？

モンテカルロ法は 1940 年代、**マンハッタン計画（Los Alamos）** において  
複雑すぎる核物理の問題を「確率的手法」で解くために開発された方法です。

**ポイント**  
- 本来は解析解の存在しない複雑な問題を  
  **ランダムサンプリング（乱数）** によって数値的に近似する手法  
- 「確率的アプローチで決定論的問題を解く」  

---

## 2. 例：円の面積をモンテカルロで求める

通常、円の面積は

$ A = \pi r^2 $

で求められますが、モンテカルロ法では以下のように求められます：

1. 正方形内（$[0, A] \times [0, A]$）に乱数で点を大量に打つ  
2. そのうち「円の内部に入った点」の比率を求める  
3.  
   $$
   \text{円の面積} \approx A^2 \times \frac{\text{円の内部の点数}}{\text{全点数}}
   $$

**πも式も不要**。  
確率と乱数だけで面積を推定できる → これがモンテカルロ法の本質。

---

## 3. 基本原理（重要）

- **不確定な事象の結果を多数生成し、その平均で近似値を得る**  
- 試行回数 $N$ が大きいほど精度が高くなる  
- 少ない試行では誤差が大きくなる（大数の法則）

---

## 4. 株価推定にも応用可能

株価は近似的に **幾何ブラウン運動（GBM）** に従う：

$$
dS(t) = \mu S(t)\,dt + \sigma S(t)\,dW(t)
$$

よって、未来の株価を推定したいなら：

1. **ランダムウォーク（GBM）を何万回もシミュレーションする**
2. **各シミュレーションの結果を平均する**

→ すると「最も確率の高い株価の推移」を推定できる。

---

## 5. イメージ図（概念）

1 回のシミュレーション：  
「1 本の株価パス」

数万回のシミュレーション：  
大量のギザギザのパスが重なる

その平均値：  
「将来株価の予測線」に相当

---

## 6. 実際の流れ（重要）

1. 初期株価 $S_0$ を決める  
2. 歴史データから $\mu$（平均リターン）、$\sigma$（ボラティリティ）を推定  
3. GBM に従って **1 回のランダムパス** を生成  
4. これを **数万回繰り返す**  
5. すべてのシミュレーションの平均を取り、株価を推定する

---

## 7. まとめ

- モンテカルロ法は「乱数」を使って複雑な問題を近似する方法  
- 核物理から金融まで幅広く使われている  
- 株価のモデル（GBM）は確率的なので相性が良い  
- 多数のパスを生成して平均することで、将来株価の推計ができる  

この理論を使って、次の章から **ブラック–ショールズモデルのモンテカルロ実装** に入ります。



#13-3 モンテカルロ法による将来株価シミュレーション

この資料では、講義で説明された  
**「幾何ブラウン運動（GBM）を用いた将来株価のシミュレーション」**  
をコード付きでわかりやすくまとめます。

---

## 1. 株価モデル：幾何ブラウン運動（GBM）

株価 $S(t)$ の動きは次の確率微分方程式（SDE）で表されます。

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

ここで：

- $\mu$：期待リターン（平均的な上昇率）
- $\sigma$：ボラティリティ（変動の大きさ）
- $W_t$：ウィーナー過程（ブラウン運動）

この SDE を Itô の補題で解くと、解析解は次になります。

$$
S(t) = S_0 \exp\left[(\mu - \frac{1}{2}\sigma^2)t + \sigma W_t\right]
$$

これを使えば、将来株価をランダムに生成できます。

---

## 2. シミュレーションの考え方

モンテカルロ法では、

- **GBM のランダムパスを何本も生成**
- **各パスの平均を取ることで、最も確率が高い経路を推定**

します。

1本のパスは次の式で生成できます。

$$
S_{t+1} = S_t \times \exp\left[(\mu - \frac12 \sigma^2) + \sigma Z\right]
$$

ただし  
$Z \sim N(0,1)$（標準正規分布）

---

## 3. Python実装：モンテカルロ株価シミュレーション

以下は、講義内容を完全に再現した実装です。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Monte Carlo Stock Price Simulation ---
NUM_SIMULATIONS = 1000

def stock_montecarlo(S0, mu, sigma, n_days, sims=NUM_SIMULATIONS):
    """
    幾何ブラウン運動による株価のモンテカルロシミュレーション
    S0 : 初期株価
    mu : 平均リターン
    sigma : ボラティリティ
    n_days : 生成する日数
    sims : シミュレーション回数
    """
    results = []

    for _ in range(sims):
        prices = [S0]

        for _ in range(n_days):
            prev_price = prices[-1]

            # 1日のGBMモデル → dt = 1 と仮定
            next_price = prev_price * np.exp(
                (mu - 0.5 * sigma**2) + sigma * np.random.normal(0, 1)
            )
            prices.append(next_price)

        results.append(prices)

    df = pd.DataFrame(results).T  # 列＝シミュレーション1本
    return df


# --- メイン処理 ---
if __name__ == "__main__":
    S0 = 50        # 初期株価
    mu = 0.00002   # 平均リターン（小さい値）
    sigma = 0.01   # ボラティリティ
    n_days = 252   # 1年分の取引日数

    sim_data = stock_montecarlo(S0, mu, sigma, n_days)

    # すべてのパスをプロット
    sim_data.plot(legend=False, figsize=(10,5), alpha=0.3)
    plt.title("Monte Carlo Simulation of Future Stock Prices")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.show()

    # 平均パスを計算
    sim_data["mean"] = sim_data.mean(axis=1)

    # 平均パスをプロット
    sim_data["mean"].plot(figsize=(10,5), color="red", linewidth=2)
    plt.title("Mean Path of Monte Carlo Simulation")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.show()

    # 1年後の株価予測
    prediction = sim_data["mean"].iloc[-1]
    print(f"Prediction for future stock price after 1 year: ${prediction:.2f}")
```
## 4. シミュレーション結果の見方

- **細い線**：多数の GBM パス（ランダムな未来を表す）
- **赤の太線（mean）**：最も確率の高い軌道（平均パス）
- **最後の値**：1年後の株価予測値（統計的な推定）

---

## 5. この方法の利点と注意点

###  利点
- 解析的に解けない問題も近似可能  
- ボラティリティを変えることでリスクシナリオを作れる  
- Black–Scholes の理論と整合性がある  

###  注意点
- 平均リターンと $\sigma$ の推定に依存する

---

- 株価は **幾何ブラウン運動（GBM）** でモデル化できる  
- その解析解を使えば **将来株価のモンテカルロシミュレーション** が可能  
- 多数のパスを生成し、平均を取ることで未来の最尤経路を推定できる  
- Python では数十行で実装可能  


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Monte Carlo Stock Price Simulation ---
NUM_SIMULATIONS = 1000

def stock_montecarlo(S0, mu, sigma, n_days, sims=NUM_SIMULATIONS):
    """
    幾何ブラウン運動による株価のモンテカルロシミュレーション
    S0 : 初期株価
    mu : 平均リターン
    sigma : ボラティリティ
    n_days : 生成する日数
    sims : シミュレーション回数
    """
    results = []

    for _ in range(sims):
        prices = [S0]

        for _ in range(n_days):
            prev_price = prices[-1]

            # 1日のGBMモデル → dt = 1 と仮定
            next_price = prev_price * np.exp(
                (mu - 0.5 * sigma**2) + sigma * np.random.normal(0, 1)
            )
            prices.append(next_price)

        results.append(prices)

    df = pd.DataFrame(results).T  # 列＝シミュレーション1本
    return df


# --- メイン処理 ---
if __name__ == "__main__":
    S0 = 50        # 初期株価
    mu = 0.00002   # 平均リターン（小さい値）
    sigma = 0.01   # ボラティリティ
    n_days = 252   # 1年分の取引日数

    sim_data = stock_montecarlo(S0, mu, sigma, n_days)

    # すべてのパスをプロット
    sim_data.plot(legend=False, figsize=(10,5), alpha=0.3)
    plt.title("Monte Carlo Simulation of Future Stock Prices")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.show()

    # 平均パスを計算
    sim_data["mean"] = sim_data.mean(axis=1)

    # 平均パスをプロット
    sim_data["mean"].plot(figsize=(10,5), color="red", linewidth=2)
    plt.title("Mean Path of Monte Carlo Simulation")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.show()

    # 1年後の株価予測
    prediction = sim_data["mean"].iloc[-1]
    print(f"Prediction for future stock price after 1 year: ${prediction:.2f}")
````

#13-4 モンテカルロ法によるオプション価格計算の準備（理論編）

前回は「将来の株価をモンテカルロ法で推定する方法」を説明した。  
今回の内容は **オプション価格のモンテカルロ計算** に進むための理論的準備となる。

---

## 1. 基本アイデア：将来の株価が必要

オプション価格は、将来の株価に依存する。

- コール：$\max(S_T - K, 0)$  
- プット：$\max(K - S_T, 0)$  

ここで $S_T$ は **満期 $T$ の株価**。

➡️ よって **Monte Carlo Simulation で $S_T$ を大量に生成する必要がある**。

---

## 2. 株価のモデル：幾何ブラウン運動（GBM）

株価は次の **確率微分方程式（SDE）** に従うと仮定：

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

この解は **指数関数（ログ正規分布）** になり：

$$
S_T = S_0 \exp\!\left[
\left( \mu - \frac12\sigma^2 \right)T
+ \sigma \sqrt{T} Z
\right]
$$

ここで $Z \sim N(0,1)$。

---

## 3. リスク中立化：$\mu$ を $r$ に置き換える

ブラック–ショールズの考え方では、

- 現実世界の平均リターン $\mu$  
  （企業成長、投資家期待など）

ではなく、

- **リスクフリーレート $r$**

を使う必要がある（リスク中立測度）。

従って、計算に使う式は：

$$
S_T = S_0 \exp\!\left[
\left( r - \frac12\sigma^2 \right)T
+ \sigma \sqrt{T} Z
\right]
$$

---

## 4. Monte Carlo でやること

満期における株価 $S_T$ を多数生成する：
```python
for i in range(N_simulations):
    Z = 標準正規乱数
    S_T[i] = S0 * exp((r - 0.5σ²)T + σ√T Z)
```

### ● コールオプション
$$
\text{Payoff}_i = \max(S_T^{(i)} - K,\; 0)
$$

### ● プットオプション
$$
\text{Payoff}_i = \max(K - S_T^{(i)},\; 0)
$$

---

## 5. 割引して現在価値にする

未来のキャッシュフローを現在価値へ割り引く：

$$
C_0 = e^{-rT} \cdot \frac{1}{N}\sum_{i=1}^N \text{Payoff}_i
$$

➡ これが Monte Carlo によるオプション価格。

---

## 6. なぜ Monte Carlo が必要か？

- ブラック–ショールズ解析解が使えないケースに対応（例：アメリカンオプション）
- ペイオフが非線形でも問題なし
- パス依存型（Asian, Barrier, Lookback など）にも適用可能  

**→ 複雑な条件付きデリバティブでは Monte Carlo が標準ツール**

---

## まとめ

- 株価は **GBM（幾何ブラウン運動）** でモデル化される  
- リスク中立測度では **平均リターン $\mu$ → 金利 $r$** に置換  
- Monte Carlo で大量の $S_T$ を生成  
- ペイオフを計算し、割引平均して **現在価値＝オプション価格** を得る  

#13-5 Monte Carlo によるコールオプション価格シミュレーション（Python実装）

この章では、講義で説明された **Monte Carlo を使ったオプション価格計算** を  
Python クラスとして実装します。

---

## 1. モデル式（GBM の解析解）

Monte Carlo では幾何ブラウン運動（GBM）の解析解を使って満期株価 $S_T$ を生成します：

$$
S_T = S_0 \exp\left( (r - 0.5\sigma^2)T + \sigma\sqrt{T}\,Z \right),
\quad Z\sim N(0,1)
$$

- $S_0$：初期株価  
- $r$：無リスク金利（リスク中立測度のドリフト）  
- $\sigma$：ボラティリティ  
- $T$：満期  
- $Z$：標準正規乱数  

---

## 2. コールオプションの payoff

コールオプションの場合：

$$
\text{Payoff} = \max(S_T - K,\;0)
$$

---

## 3. 現在価値へ割引

未来の payoff を現在価値に変換：

$$
C_0 = e^{-rT} \cdot \mathbb{E}[\text{Payoff}]
$$

Monte Carlo では  
**多数のパスを生成 → payoff の平均 → 割引**  
で求める。

---

## 4. Python 実装（クラスベース）

```python
import numpy as np

class OptionPricingMonteCarlo:

    def __init__(self, S0, K, T, r, sigma, iterations):
        """
        Monte Carlo 法によるオプション価格計算クラス
        S0 : 初期株価
        K  : 標準価格（ストライク）
        T  : 満期（年）
        r  : 無リスク金利
        sigma : ボラティリティ
        iterations : シミュレーション回数
        """
        self.S0 = S0
        self.K = K
        self.T = T
        self.r = r
        self.sigma = sigma
        self.iterations = iterations

    def call_option_simulation(self):
        """
        Monte Carlo によるコールオプション価格の近似
        """

        # --- 1. 標準正規乱数（Z） ---
        Z = np.random.normal(0, 1, self.iterations)

        # --- 2. GBM による満期株価 S_T ---

        payoff_mean = payoff.mean()

        # --- 5. 割引（現在価値） ---
        call_price = np.exp(-self.r * self.T) * payoff_mean

        return call_price
```

#13-6 Monte Carlo によるプットオプション価格計算

前回は **コールオプション価格** を Monte Carlo 法で計算したが、  
今回は **プットオプション価格** を同じ要領で計算する。

---

## 1. コールとプットの違い

コールオプション  
$$\text{Payoff} = \max(S_T - K,\;0)$$

プットオプション  
$$\text{Payoff} = \max(K - S_T,\;0)$$

違いは **$S_T - K$ を使うか、$K - S_T$ を使うか** のみ。

---

## 2. Python 実装（クラス構造）

下記は前回のクラスをベースにし、  
**put_option_simulation()** を追加した完全版コード。

```python
import numpy as np

class OptionPricing:
    def __init__(self, S0, K, T, r, sigma, iterations):
        self.S0 = S0          # 初期株価
        self.K = K            # ストライク
        self.T = T            # 満期
        self.r = r            # 無リスク金利
        self.sigma = sigma    # ボラティリティ

    def call_option_simulation(self):

        payoff = np.maximum(self.K - S_T, 0)
```

##4 まとめ

- コールとプットの違いは **ペイオフ関数のみ**
- GBM（幾何ブラウン運動）の株価モデル式：

$$
S_T = S_0 \exp\left[(r - \tfrac{1}{2}\sigma^2)T + \sigma\sqrt{T}\,Z\right]
$$

- Monte Carlo では大量の $S_T$ を生成  
- ペイオフを平均し、最後に **$e^{-rT}$ で割り引く → 現在価値 = オプション価格**


````python
import numpy as np

class OptionPricing:
    def __init__(self, S0, K, T, r, sigma, iterations):
        self.S0 = S0          # 初期株価
        self.K = K            # ストライク
        self.T = T            # 満期
        self.r = r            # 無リスク金利
        self.sigma = sigma    # ボラティリティ
        self.iter = iterations

    # --- コールオプション ---
    def call_option_simulation(self):
        # 標準正規乱数
        Z = np.random.normal(0, 1, self.iter)

        # GBM による S_T の生成
        S_T = self.S0 * np.exp(
            (self.r - 0.5 * self.sigma**2) * self.T
            + self.sigma * np.sqrt(self.T) * Z
        )

        # コールのペイオフ
        payoff = np.maximum(S_T - self.K, 0)

        # 割引して現在価値へ
        price = np.exp(-self.r * self.T) * np.mean(payoff)
        return price

    # --- プットオプション ---
    def put_option_simulation(self):
        # 標準正規乱数
        Z = np.random.normal(0, 1, self.iter)

        # 将来株価 S_T を生成
        S_T = self.S0 * np.exp(
            (self.r - 0.5 * self.sigma**2) * self.T
            + self.sigma * np.sqrt(self.T) * Z
        )

        # プットのペイオフ（唯一の違い）
        payoff = np.maximum(self.K - S_T, 0)

        # 現在価値に割引
        price = np.exp(-self.r * self.T) * np.mean(payoff)
        return price


# --- メイン処理 ---
if __name__ == "__main__":
    model = OptionPricing(
        S0=100,     # 初期株価
        K=100,      # ストライク
        T=1,        # 満期1年
        r=0.05,     # 無リスク金利 5%
        sigma=0.2,  # ボラティリティ
        iterations=10000  # シミュレーション回数
    )

    call_price = model.call_option_simulation()
    put_price = model.put_option_simulation()

    print(f"Call Option Price  = ${call_price:.2f}")
    print(f"Put Option Price   = ${put_price:.2f}")
````
